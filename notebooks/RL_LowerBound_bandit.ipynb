## バンディットの下界の出し方（特殊）

* 行動空間：$\mathcal{A}:=\{1, \ldots, A\}^d$
* アーム $i$はパラメータ$p_i$のベルヌーイ分布に従う．$r_i \in \{0, 1\}$とする．

次のバンディットを考える：
* $\mathcal{B}_0$：$p_i=0.5$ for all $i$
* $\mathcal{B}_i$ for $(i=1, \dots, A)$：$p_i=0.5 + \Delta$ and $p_l=0.5$ for $i\neq j$

$\Delta$は後で決めるパラメータ

* $\mathbb{P}_i$はアルゴリズムを$\mathcal{B}_i$で走らせたときの確率測度
* $\mathbb{P}'_i$は$(a_k, r_k)_{k=1}^K$ がアルゴリズムで誘導される確率測度とする．

このとき，次が成立します．

---

参考：[Adapting to game trees in zero-sum imperfect information games](https://proceedings.mlr.press/v202/fiegel23a/fiegel23a.pdf#page=15.08)

固定された$\rho \in (0, 1)$と$v \in (0, \infty)$を考えましょう．
* $s:=\frac{4 \Delta}{3} \log \frac{1}{\rho}+\sqrt{2 v \log \frac{1}{\rho}}$ および
* $n_i:=\mathbb{E}_0\left[\sum_{k=1}^K \mathbb{I}_{\left\{a_k=i\right\}}\right]$とします．

このとき，$0 < \Delta \leq 0.3$について，

$$
\mathbb{P}_i\left(f_i\left(H_K\right)=1\right)
\geq 
\exp (-s-v)\left(\mathbb{P}_0\left(f_i\left(H_K\right)=1\right)-\rho-\frac{4\Delta^2}{v}n_i\right)
$$

が，任意の関数
$f_i:([A] \times\{0,1\})^K \rightarrow\{0,1\}$で成立します．ここで，$H_K=\left(a_k, r_k\right)_{k=1}^K$および
$S_i=\sum_{k=1}^K \mathbb{I}_{\left\{a_k=i\right\}}\left[r_k \ln (1+2 \Delta)+\left(1-r_k\right) \ln (1-2 \Delta)\right]$としました．


---

**証明**

まず，
$$
\mathbb{P}_i\left(f_i\left(H_K\right)=1\right)=\mathbb{E}_i\left[f_i\left(H_K\right)\right]=\mathbb{E}_i^{\prime}\left[f_i\left(H_K\right)\right]=\mathbb{E}_0^{\prime}\left[f_i\left(H_K\right) \frac{d \mathbb{P}_i^{\prime}}{d \mathbb{P}_0^{\prime}}\left(H_K\right)\right]
$$

が成立します．ここで，$i$と$0$は$p_i$しか変わらない＆$r_k \in \{0, 1\}$のため，
$$
\begin{aligned}
\frac{d \mathbb{P}_i^{\prime}}{d \mathbb{P}_0^{\prime}}\left(H_K\right) 
& =\prod_{k=1}^K\left[r_k \frac{0.5+\Delta \mathbb{I}_{\left\{a_k=i\right\}}}{0.5}+\left(1-r_k\right) \frac{0.5-\Delta \mathbb{I}_{\left\{a_k=i\right\}}}{0.5}\right] \\
& =\prod_{k=1}^K \left[r_k (1+2\Delta \mathbb{I}_{\left\{a_k=i\right\}})+\left(1-r_k\right) \left(1-2 \Delta \mathbb{I}_{\left\{a_k=i\right\}}\right)\right]\\
& =\prod_{k=1}^K (1 + x_k)
% & =\exp\left(\sum_{k=1}^K \ln \left[r_k (1+2\Delta \mathbb{I}_{\left\{a_k=i\right\}})+\left(1-r_k\right) \left(1-2 \Delta \mathbb{I}_{\left\{a_k=i\right\}}\right)\right]\right) \\
% & =\exp \{\underbrace{\sum_{k=1}^K \mathbb{I}_{\left\{a_k=i\right\}}\left[r_k \ln (1+2 \Delta)+\left(1-r_k\right) \ln (1-2 \Delta)\right]}_{=S_i}\}
\end{aligned}
$$

が成立します．
ここで，$x_k$は$2 \Delta\left(2 r_k-1\right) \mathbb{I}_{\left\{a_k=i\right\}}$としました．
$0 \leq \Delta \leq 0.3$なら，$x_k \geq -0.6$が成立します．

$(1+x) \geq \exp(x - x^2)$が$x\geq -0.6$で成立するので，
$S_K:=-\sum_{k=1}^K x_k$ and $V_K:=\sum_{k=1}^T x_k^2=4 \Delta^2 \sum_{k=1}^K \mathbb{I}_{\left\{a_k=i\right\}}$について，

$$
\frac{d \mathbb{P}_i^{\prime}}{d \mathbb{P}_0^{\prime}}\left(H_K\right) 
\geq \exp \left(-S_K-V_K\right)
$$

が成り立ちます．
よって，任意の$s, v \in \mathbb{R}$について，

$$
\begin{aligned}
\mathbb{P}_i\left(f_i\left(H_K\right)=1\right)
&\geq 
\mathbb{E}_0^{\prime}\left[f_i\left(H_K\right) \frac{d \mathbb{P}_i^{\prime}}{d \mathbb{P}_0^{\prime}}\left(H_K\right)\right]\\
&\geq \mathbb{E}_0^{\prime}\left[f_i\left(H_K\right)\exp(-S_k - V_K)\right]\\
&\geq \mathbb{E}_0^{\prime}\left[f_i\left(H_K\right)\mathbb{I}_{\{S_K < s, V_K \leq v\}}\exp(-S_k - V_K)\right]\\
&\geq \exp(-s - v)\mathbb{E}_0\left[f_i\left(H_K\right)\mathbb{I}_{\{S_K < s, V_K \leq v\}}\right]\\
\end{aligned}
$$

最後に，$f_i\left(H_K\right)\mathbb{I}_{\{S_K < s, V_K \leq v\}}$を下から抑えましょう．
$A \cap B=\left(A \cup B^c\right) \cap B$を使うと，$f$の出力は$0$か$1$なので，

$$
\begin{aligned}
f\left(H_{K+1}\right) \mathbb{I}_{\left\{S_K<s, V_K \leq v\right\}} & =f\left(H_{K+1}\right)\left(1-\mathbb{I}_{\left\{S_K \geq s, V_K \leq v\right\}}-\mathbb{I}_{\left\{V_K>v\right\}}\right) \\
& \geq f\left(H_{K+1}\right)-\mathbb{I}_{\left\{S_K \geq s, V_K \leq v\right\}}-\mathbb{I}_{\left\{V_K>v\right\}},
\end{aligned}
$$

が成立します．よって，
$$
\begin{aligned}
\mathbb{E}_0\left[
f\left(H_{K+1}\right)\mathbb{I}_{\left\{S_K<s, V_K \leq v\right\}}\right] 
& =\mathbb{E}_0[f\left(H_{K+1}\right)] 
-\mathbb{P}_0 \left(S_K \geq s, V_K \leq v\right)
-\mathbb{P}_0 \left(V_K>v\right),
\end{aligned}
$$

最後に，２項目と３項目をバウンドしましょう．
３項目は，マルコフの不等式を使うと，上から
$\mathbb{P}_0 \left(V_K>v\right) \leq \mathbb{E}_0\left[V_K\right] / v$とバウンドできます．

２項目はfreedmanの不等式でバウンドします．

* $F_k'$を$(s_h, r_h)_{h=1}^k$で生成された$\sigma$-代数とします．
* $|x_k|$はalmost surelyで$2\Delta$以下です
* $\{x_k\}_{k=1}^K$は$F_k'$上のマルチンゲール差分列です．
* $\sum_{k=1}^K \mathbb{E}_0\left[\left(x_k-\mathbb{E}_0\left[x_k \mid \mathcal{F}_k^{\prime}\right]\right)^2 \mid \mathcal{F}_k\right]=V_K$が$\mathbb{E}_0\left[x_k\mid F_k'\right]=0$なので成立してます．

以上から，freedmanの不等式（Bernsteinの不等式とほぼ同じですが，Martingaleで使えます）より，
$$
s:=\frac{4 \Delta}{3} \log \frac{1}{\rho}+\sqrt{2 v \log \frac{1}{\rho}}
$$
とすると，$\mathbb{P}_0 \left(S_K \geq s, V_K \leq v\right) \leq \rho$が成り立ちます．

よって，この$s$について，

$$
\begin{aligned}
\mathbb{P}_i\left(f_i\left(H_K\right)=1\right)
&\geq 
\exp (-s-v)\left(\mathbb{P}_0\left(f_i\left(H_K\right)=1\right)-\rho-\frac{4\Delta^2}{v}\mathbb{E}_0\left[\sum^K_{k=1} \mathbb{I}_{a_k = i}\right]\right)\\
&=
\exp (-s-v)\left(\mathbb{P}_0\left(f_i\left(H_K\right)=1\right)-\rho-\frac{4\Delta^2}{v}n_i\right)\\
\end{aligned}
$$

が成立します．

## リグレットの下界

これを使うと，リグレットの下界がすぐに出せます．

$E_i$を，アルゴリズムが$i$を選ぶ総確率が$0.5$以下の事象とします:
$$
E_i=\left\{\sum_{k=1}^K \pi_k(i) \leq 0.5 K\right\}
$$

ここで，アルゴリズムは履歴から$\pi_k$を決定的に吐き出すとします（確率的な場合への拡張は容易です）．
このとき，$\mathbb{I}_{E_i}$を$f_i$として書けます．よって，上の補題を使うと，

$$
\max _{i \in[A]} \mathbb{P}_i\left(E_i\right) \geq \frac{1}{A} \sum_{i \in[A]} \mathbb{P}_i\left(E_i\right)>\exp (-s-v)\left(\frac{1}{A} \sum_{i=1}^A \mathbb{P}_0\left(E_i\right)-\rho-\frac{4 K \Delta^2}{A v}\right)
$$

ここで，インスタンス$\mathcal{B}_0$にて，$\mathbb{I}_{\left\{E_i\right\}}=0$ が何らかの$i$で成り立つとき，$i$を半数以上で選択しているので，他の$i \neq j$では選択していないことになります．つまり，$\mathbb{I}_{\left\{E_j\right\}}=1$ for $j \neq i$ なので，
$$\sum_{i \in[A]} \mathbb{P}_0\left(E_i\right)=\mathbb{E}_0\left[\sum_{i \in[A]} \mathbb{I}_{\left\{E_i\right\}}\right] \geq A-1$$
が成立します．よって，

$$
\max _{i \in[A]} \mathbb{P}_i\left(E_i\right)>\exp (-s-v)\left(1-\frac{1}{A}-\rho-\frac{4 K \Delta^2}{A v}\right) .
$$

が成り立ちます．このとき，
$$
v=\frac{32 K \Delta^2}{A} \text { and } \rho=\frac{1}{8}
$$
とすると，
$$
\left(1-\frac{1}{A}-\rho-\frac{4 K \Delta^2}{A v}\right)
= \frac{3}{4} - \frac{1}{A}\geq 1/4
$$
および
$$
\begin{aligned}
-s-v
&=-\frac{4 \Delta}{3} \log\frac{1}{\rho} -\sqrt{2 v \log \frac{1}{\rho}} -v \\
&=-\frac{4 \Delta}{3} \log 8-\sqrt{\frac{64 K \Delta^2}{A} \log 8} -\frac{32 K \Delta^2}{A} \\
&=-4\Delta\log 2- 8\Delta \sqrt{\frac{3K}{A} \log 2} -\frac{32 K \Delta^2}{A} \\
\end{aligned}
$$
なので，
$$
\max _{i \in[A]} \mathbb{P}_i\left(E_i\right)>\frac{1}{4} \exp \left(-4 \Delta \log 2-8 \Delta \sqrt{\frac{3 K}{A} \log 2}-\frac{32 K \Delta^2}{A}\right)
=: \delta
$$
が成立します（$A \geq 2$が必要です）．

ここで，$\delta$を使って$\Delta$を変形すると，
$$
\Delta=\sqrt{\left(\frac{A}{16 K} \log 2+\frac{1}{8} \sqrt{\frac{3 A}{K} \log 2}\right)^2+\frac{A}{32 K} \log \frac{1}{4 \delta}}-\frac{A}{16 K} \log 2-\frac{1}{8} \sqrt{\frac{3 A}{K} \log 2}<\sqrt{\frac{A}{32 K} \log \frac{1}{4 \delta}}
$$
を得ます．$\Delta < 0.3$にするためには，$K \geq-0.4 A \log (4 \delta)$を仮定すれば問題ありません．

すなわち，このような$\Delta$のとき，
$$\max _{i \in[A]} \mathbb{P}_i\left(E_i\right)>\delta$$

が成り立ちます．
ところで，$E_i$のせいで，何らかの$\mathcal{B}_i$ではリグレットが$0.5 \Delta K$以上になるので，後は$\Delta$を代入すれば，$\sqrt{K}$の下界が出ます．

## サンプル効率の下界

サンプル効率の下界もすぐに出せます．

$E_i$を，$K$ステップ目でアルゴリズムが$i$を選ぶ確率が$0.5$以下の事象とします:
$$
E_i=\left\{\pi_K(i) \leq 0.5\right\}
$$

リグレットと全く同様にして，
$$
\max _{i \in[A]} \mathbb{P}_i\left(E_i\right)>\frac{1}{4} \exp \left(-4 \Delta \log 2-8 \Delta \sqrt{\frac{3 K}{A} \log 2}-\frac{32 K \Delta^2}{A}\right)
=: \delta
$$
が成立し，これを使って$\Delta=...$の式が出せます．すると，$K$ステップ目のリグレットは$0.25 \Delta$になり，その下界は
$\Omega(\sqrt{A/K\log (1/4\delta)})$です．よって，$K \geq A/\varepsilon^2 \log (1/4\delta)$以上のステップを経ないと，$\varepsilon$最適方策は得られません．
