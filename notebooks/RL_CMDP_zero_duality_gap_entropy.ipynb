# エントロピー正則化された制約付きMDPの双対ギャップはゼロ

参考：
* [Safe Policies for Reinforcement Learning via Primal-Dual Methods](https://arxiv.org/abs/1911.09101)
* [A Dual Approach to Constrained Markov Decision Processes with Entropy Regularization](https://arxiv.org/abs/2110.08923)


[RL_CMDP_zero_duality_gap.ipynb](RL_CMDP_zero_duality_gap.ipynb)の別のパターンの証明です．
また，こちらを使うとエントロピー正則化されたCMDPの双対性が証明できるらしいです（２番目の論文にその旨が書いてありますが，証明はありませんね）．やってみましょう．

次のCMDPを考えましょう．

$$
\begin{aligned}
& P^{\star} \triangleq \max _{\pi \in \mathcal{P}(\mathcal{S})} \quad V^\tau_0(\pi) \triangleq \mathbb{E}_{s, \pi}\left[\sum_{t=0}^{\infty} \gamma^t r_0\left(s_t, \pi\left(s_t\right)\right)\right] 
+ \tau
\mathbb{E}_{s, \pi}\left[\sum_{t=0}^{\infty} -\gamma^t \log \pi \left(a_t\mid s_t\right)\right]
\\
& \quad \text { subject to } \quad V_i(\pi) \triangleq \mathbb{E}_{s, \pi}\left[\sum_{t=0}^{\infty} \gamma^t r_i\left(s_t, \pi\left(s_t\right)\right)\right] \geq c_i, i=1, \ldots, m,
\end{aligned}
$$

ちなみに，このようなエントロピーの足し方はベルマン作用素にエントロピーを足すのと同じです（[A Theory of Regularized Markov Decision Processes](https://arxiv.org/abs/1901.11275)のDefinition２あたりを参照）．

双対ギャップがゼロになることを証明するために，次のPerturbation関数を定義します．

$$
\begin{aligned}
P(\xi) \triangleq & \max _{\pi \in \mathcal{P}(\mathcal{S})} V^\tau_0(\pi) \\
& \text { subject to } U_i(\pi) \geq c_i+\xi_i, i=1 \ldots m .
\end{aligned}
$$

まず，次を示します．

---

$r(s, a)$がboundedであり，CMDPに実行可能解が存在するとします．このとき，次の集合は空ではなく，また，凸です．

$$
\begin{aligned}
\mathcal{C}=\left\{\xi \in \mathbb{R}^{m+1} \mid \exists \pi \in \mathcal{P}_s(\mathcal{A}), V^\tau_0(\pi)\right. & \geq \xi_0, \\
U_i(\pi) \geq c_i+\xi_i, \forall i & =1, \ldots, m\} .
\end{aligned}
$$

まず，$\pi^*$は明らかに制約を満たすので，$\xi_0=P^\star$とすれば，

$$\left(P^{\star}, \mathbf{0}_m^{\top}\right)^{\top} \in \mathcal{C}$$

です．よって$C$は空ではありません．

ここで，
$\xi^1, \xi^2 \in \mathcal{C}$ および，それを達成する方策を $\pi^1, \pi^2 \in \mathcal{P}_s(\mathcal{A})$とします．
$\mu \xi^1+(1-\mu) \xi^2 \in \mathcal{C}$が任意の$\mu\in (0, 1)$について成立することを示しましょう．


まず，価値関数と制約の関数は次で表せます．$\rho$は$\pi$についてのoccupancy measureです．

$$
(1-\gamma) V^\tau_0(\pi)=\int_{\mathcal{S} \times \mathcal{A}} \left( r(s, a) - \tau \log \pi(a\mid s)\right)\rho_\pi(s, a) d s d a
$$

$$
(1-\gamma) U_i(\pi)=\int_{\mathcal{S} \times \mathcal{A}} r(s, a) \rho_\pi(s, a) d s d a
$$

また，$\mathcal{R}$を全てのOccupancy measureの集合としましょう．これは凸でありコンパクトであることは[RL_occupancy_measure.ipynb](RL_occupancy_measure.ipynb)で示しました．

ここで，$\rho^1, \rho^2 \in \mathcal{R}$を，$\pi^1$と$\pi^2$に割り当てられたOccupancy measureとします．
$\mathcal{R}$は凸なので，
そのoccupancy measureが$\mu \rho^1+(1-\mu) \rho^2$であるような方策$\pi^\mu$が存在します．
このとき，
$$
U_i\left(\pi^\mu\right) \geq \mu \xi_i^1+(1-\mu) \xi_i^2
$$

は明らかに成立します（Occupancy measureの線型性で成立します）．

一方で，
$$
\begin{aligned}
(1-\gamma) V^\tau_0(\pi^\mu)
=&\int_{\mathcal{S} \times \mathcal{A}} \left( r(s, a) - \tau \log \pi^\mu(a\mid s)\right)\rho_{\pi^\mu}(s, a) d s d a\\
=&
\mu\int_{\mathcal{S} \times \mathcal{A}} r(s, a) \rho^1(s, a)d s d a\\
&+(1-\mu)\int_{\mathcal{S} \times \mathcal{A}} r(s, a) \rho^2(s, a)d s d a\\
&-\tau\underbrace{\int_{\mathcal{S} \times \mathcal{A}} \log \pi^\mu(a\mid s) \rho^\mu(s, a)d s d a}_{(a)}\\
\end{aligned}
$$
です．(a)の項は$\rho$について凸になります（これは[A unified view of entropy-regularized Markov decision processes](https://arxiv.org/abs/1705.07798)のProposition１が参考になるはず．これは意外と自明ではないので，Entropy CMDPの論文は微妙では？）．

あとは自明です．
