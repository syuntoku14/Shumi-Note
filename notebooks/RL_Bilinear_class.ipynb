# Bilinear Class

[Bellman Rank](RL_Bellman_rank.ipynb)の続きです．
Bilinear Classでは，Q-Bellman rankとV-Bellman rankの一般化を考えます．


---

次の
* MDP
* 仮説集合$\mathcal{H}$
* 各$f\in \mathcal{H}$に定義された$\ell_f:\mathcal{S}\times \mathcal{A}\times \mathcal{S} \times[H] \times \mathcal{H}$．これはLoss関数みたいなもの？例えばベルマン誤差を使いますが，Linear Mixture MDPなどでは違う関数になります．
* Non-stationaryな方策の集合$\Pi_{e s t}=\left\{\pi_{e s t}(f): f \in \mathcal{H}\right\}$

を考えます．このとき，$\left\{\mathcal{H}, \ell_f, \Pi_{e s t}, \mathcal{M}\right\}$のタプルは次を満たすときに$d$次元のBilinear rankであるといいます．

* $\mathcal{H}$が実現可能である
* 次を満たす$W_h: \mathcal{H} \mapsto \mathbb{R}^d$および$X_h: \mathcal{H} \mapsto \mathbb{R}^d$が存在する：

任意の$f \in \mathcal{H}$と$h\in [H]$について，

$$
\left|\mathbb{E}_{s, a \sim d_h^{\pi_f}{ }_{, s^{\prime} \sim P_h(\cdot \mid s, a)}}\left[Q_{h, f}(s, a)-r(s, a)-V_{h+1, f}\left(s^{\prime}\right)\right]\right| \leq\left|\left\langle W_h(f)-W_h\left(f^{\star}\right), X_h(f)\right\rangle\right|
$$

および

$$
\left|\mathbb{E}_{s \sim d_h^{\pi_f}, a \sim \pi_{e s t, h}(s ; f), s^{\prime} \sim P_h(\cdot \mid s, a)}\left[\ell_f\left(s, a, s^{\prime}, h, g\right)\right]\right|=\left|\left\langle W_h(g)-W_h\left(f^{\star}\right), X_h(f)\right\rangle\right|,
$$

が成立する．


ここで，$\pi_{est}(f)$は一様分布か$\pi_f$そのものであることが多いです．$\pi_f$の場合はon-policyに相当します．
また，アルゴリズムが$W$と$X$を知る必要はありません．

---

**直感**

１つめは簡単です．$\pi_f$と$f$による価値のベルマン誤差を，仮説$f$によるBilinearな形式で抑えられることを表しています．

２つめがかなり大切です．
２つめの左側の$\mathbb{E}_{s \sim d_h^{\pi_f}, a \sim \pi_{e s t, h}(s ; f), s^{\prime} \sim P_h(\cdot \mid s, a)}\left[\ell_f\left(s, a, s^{\prime}, h, g\right)\right]$を見ると，$s, a, s'$のデータはすべて関数$f$から出ていることに注意しましょう．
つまり，右辺において，異なる$g$についての$\left|\left\langle W_h(g)-W_h\left(f^{\star}\right), X_h(f)\right\rangle\right|$を評価する際，
データを変化させる必要はありません．
これは教師あり学習に置ける一様収束の保証の話に似てます（TODO:）




---

**LQRの例**

LQRはBilinear classに属します．

表記：

* $\mathcal{S} \subset \mathbb{R}^d$
* $\mathcal{A} \subset \mathbb{R}^K$

次を満たすとき，MDP$\mathcal{M}$は有限ホライゾン離散時間LQRである，とします．

$$
\begin{aligned}
s_{h+1} & =A s_h+B a_h+\epsilon_h \\
r_h & =s_h^{\top} Q s_h+a_h^{\top} a_h+\tau_h
\end{aligned}
$$

を満たす
* $A \in \mathbb{R}^{d\times d}$
* $B \in \mathbb{R}^{d\times K}$
* $Q \in \mathbb{R}^{d\times d}$

が存在する．
ここで，
* $\epsilon_h$は$\mathbb{E}\left[\epsilon_h \epsilon_h^{\top}\right]=\Sigma$を満たす平均ゼロのノイズ
* $\tau_h$は$\mathbb{E}\left[\tau_h^2\right]=\sigma^2$を満たす平均ゼロのノイズ

です．

LQRがBilinear Classに属することを示すために，次を使います：

---

**補題**

$C_{\pi, h}\in \mathbb{R}^{K\times d}$を使った
Non-stationaryな線形方策
$\pi_h\left(s_h\right)=C_{\pi, h} x$
を考えると，

$$
V_h^\pi\left(s_h\right)=s_h^{\top} \Lambda_{\pi, h} s_h+O_{\pi, h} \text { for some } \Lambda_{\pi, h} \in \mathbb{R}^{d \times d} \text { and } O_{\pi, h} \in \mathbb{R} \text {. }
$$

が成立する．

証明は
[Contextual Decision Processes with Low Bellman Rank are PAC-Learnable](https://arxiv.org/abs/1610.09512)のLemma7などを参照してください．

---

ここで，仮説集合として

$$
\mathcal{H}_h=\left\{\left(C_h, \Lambda_h, O_h\right): C_h \in \mathbb{R}^{K \times d}, \Lambda_h \in \mathbb{R}^{d \times d}, O_h \in \mathbb{R}\right\}
$$

を考え，仮説$f \in \mathcal{H}$に対して

$$
\pi_f\left(s_h\right)=C_{h, f} s_h, \quad V_{h, f}\left(s_h\right)=s_h^{\top} \Lambda_{h, f} s_h+O_{h, f}
$$

とします．
また，任意の仮説$g\in \mathcal{H}$に対して，$\ell_f$を

$$
\begin{aligned}
\ell_f\left(o_h, g\right) & =Q_{h, g}\left(s_h, a_h\right)-r_h-V_{h+1, g}\left(s_{h+1}\right) \\
& =s_h^{\top} \Lambda_{h, g} s_h+O_{h, g}-s_h^{\top} Q s_h-s_h^{\top} C_{h, g}^{\top} C_{h, g} s_h-\tau_h-s_{h+1}^{\top} \Lambda_{h+1, g} s_{h+1}-O_{h+1, g} .
\end{aligned}
$$

とします（ここで$o_h=\left(r_h, s_h, a_h, s_{h+1}\right)$です）．

最後に$\pi_{e s t}(f)=\pi_f$ for $f \in \mathcal{H}$とします．

このとき，$
\left(\mathcal{H}, \ell_f, \Pi_{\mathrm{est}}, \mathcal{M}\right)
$はBilinear Classになります．

**証明**

$g=f$のとき，

$$
\mathbb{E}_{s_h, a_h, s_{h+1} \sim d^\pi f}\left[\ell\left(o_h, f\right)\right]=\mathbb{E}_{s_h, a_h, s_{h+1} \sim d^{\pi_f}}\left[Q_{h, f}\left(s_h, a_h\right)-r\left(s_h, a_h\right)-V_{h+1, f}\left(s_{h+1}\right)\right]
$$

が成り立つことに注意しましょう．Bilinear Classの条件を思い出すと，今回は２つ目のやつを示せば十分です．

次が成り立つことに注意しましょう．

$$
\begin{aligned}
& \mathbb{E}_{a_h, s_h, s_{h+1} \sim d^{\pi_f}}\left[\ell_f\left(o_h, g\right)\right] \\
&= \mathbb{E}_{s_h, s_{h+1} \sim d^{\pi_f}}\left[s_h^{\top} \Lambda_{h, g} s_h+O_{h, g}-s_h^{\top} Q s_h-s_h^{\top} C_{h, g}^{\top} C_{h, g} s_h-s_{h+1}^{\top} \Lambda_{h+1, g} s_{h+1}-O_{h+1, g}\right] \\
&= \operatorname{trace}\left(\left(\Lambda_{h, g}-Q-C_{h, g}^{\top} C_{h, g}-\left(A+B C_{h, g}\right)^{\top} \Lambda_{h+1, g}\left(A+B C_{h, g}\right)\right) \mathbb{E}_{s_h \sim d^{\pi_f}}\left[s_h s_h^{\top}\right]\right) \\
&-\operatorname{trace}\left(\Lambda_{h+1, g} \Sigma\right)+O_{h, g}-O_{h+1, g} \\
&=\left\langle W_h(g)-W_h\left(f^{\star}\right), X_h(f)\right\rangle
\end{aligned}
$$

であり，ここで，

$$
\begin{aligned}
X_h(f):= & {\left[\operatorname{vec}\left(\mathbb{E}_{s_h \sim d^{\pi_f}}\left[s_h s_h^{\top}\right]\right), 1\right], } \\
W_h(g):= & {\left[\operatorname{vec}\left(\Lambda_{h, g}-Q-C_{h, g}^{\top} C_{h, g}-\left(A+B C_{h, g}\right)^{\top} \Lambda_{h+1, g}\left(A+B C_{h, g}\right)\right),\right.} \\
& \left.O_{h, g}-O_{h+1, g}-\operatorname{trace}\left(\Lambda_{h+1, g} \Sigma\right)\right] .
\end{aligned}
$$

です．ここで，$\left\langle W_h\left(f^{\star}\right), X_h(f)\right\rangle=0$がベルマン方程式から成立することを使っています．つまり，$a_h=C_{h, f^\star} s_h$について，

$$
\begin{aligned}
& s_h^{\top} \Lambda_{h, f^{\star}} s_h+O_{h, f^{\star}}-\mathbb{E}_{s_{h+1} \sim P\left(s_h, a_h\right)}\left[s_{h+1}^{\top} \Lambda_{h+1, f^{\star}} s_{h+1}\right]-O_{h+1, f^{\star}}-s_h^{\top} Q s_h \\
& -s_h^{\top} C_{h, f^{\star}}^{\top} C_{h, f^{\star}} s_h=0 \\
\Longrightarrow & s_h^{\top} \Lambda_{h, f^{\star}} s_h+O_{h, f^{\star}}-\operatorname{trace}\left(\left[\left(A+B C_{h, f^{\star}}\right)^{\top} \Lambda_{h+1, f^{\star}}\left(A+B C_{h, f^{\star}}\right)\right] s_h s_h^{\top}\right) \\
& -\operatorname{trace}\left(\Lambda_{h+1, f^{\star}} \Sigma\right)-O_{h+1, f^{\star}}-s_h^{\top} Q s_h-\operatorname{trace}\left(C_{h, f^{\star}} C_{h, f^{\star}} s_h s_h^{\top}\right)=0 \\
\Longrightarrow & \operatorname{trace}\left(\left(\Lambda_{h, f^{\star}}-Q-C_{h, f^{\star}} C_{h, f^{\star}}-\left(A+B C_{h, f^{\star}}\right)^{\top} \Lambda_{h+1, f^{\star}}\left(A+B C_{h, f^{\star}}\right)\right) s_h s_h^{\top}\right) \\
& +O_{h, f^{\star}}-O_{h+1, f^{\star}}-\operatorname{trace}\left(\Lambda_{h+1, f^{\star}} \Sigma\right)=0
\end{aligned}
$$

が成り立ちます．最後に$s_h \sim d^{\pi_f}$で期待値を取れば終わりです．

## アルゴリズム

次のアルゴリズムを考えます．

![BLin-UCB](figs/BLin-UCB.png)

3行目の制約付き最適化が一番大事な部分です．
ここでは，すべての$h \in [0, \dots, H-1]$について，仮説集合が現在のバッチデータに対してconsistentなものになるように制約を課しています．
これによって$f^\star$ではないものを排除します．
また，$R$は$f^\star$が常に高い確率で実行可能解になるように選択します．

---

**$\pi_f$と$\pi_{est}$の違いについて**

$\pi_{e s t}=\pi_{f_t}$の場合，長さ$H$の軌跡を$m$個生成すれば大丈夫です．

そうでない場合は，それぞれの$h \in [H]$について軌跡を生成しなければなりません．
それぞれの$h \in [H]$について$\pi_{f_t}$をまず走らせて$s_h$を生成し，そのあとで$a_h \sim \pi_{e s t}$を生成する必要があります．
よって，結局$Hm$個の軌跡が必要です．


## サンプル効率の導出

まず，次を仮定します．

---

**仮定**

任意の分布
$\nu \in \Delta(\mathcal{S} \times \mathcal{A} \times \mathcal{S})$, $f \in \mathcal{H}$, そして$\delta \in(0,1)$に対して，確率$1-\delta$以上で，i.i.d.なサンプル$\mathcal{D}:=\left\{s_i, a_i, s_i^{\prime}\right\}_{i=1}^m \sim \nu$
に対して，

$$
\sup _{g \in \mathcal{H}}\left|\mathbb{E}_{\mathcal{D}}\left[\ell_f\left(s, a, s^{\prime}, h, g\right)\right]-\mathbb{E}_\nu\left[\ell_f\left(s, a, s^{\prime}, h, g\right)\right]\right| \leq \varepsilon_{g e n}(m, \mathcal{H}, \delta)
$$

が成り立つような関数$\varepsilon_{\text {gen }}(m, \mathcal{H}, \delta)$が存在する．
（ここで，$\mathbb{E}_{\mathcal{D}}$は$\mathcal{D}$上の一様分布に対する期待値を表す）

---

これはよくある一様収束の話ですね．多くの場合，$\varepsilon_{\text {gen }}(m, \mathcal{H}, \delta) \rightarrow 0, m \rightarrow \infty$であるような関数を選択できます．

例えば$\mathcal{H}$が有限であり，$\ell_f$としてQ-Bellman rankのLossを選択する場合を考えましょう．つまり，
$$
\ell_f(s_h, a_h, s_{h+1}, h, g) = 
Q_{h, g}\left(s_h, a_h\right)-r\left(s_h, a_h\right)-V_{h+1, g}\left(s_{h+1}\right)
$$
を選択します．このとき，Hoeffdingの不等式から，

$$
\sup _{g \in \mathcal{H}}\left|\mathbb{E}_{\mathcal{D}}\left[\ell_f\left(s, a, s^{\prime}, h, g\right)\right]-\mathbb{E}_\nu\left[\ell_f\left(s, a, s^{\prime}, h, g\right)\right]\right| \leq 2 H \sqrt{\ln (|\mathcal{H}| / \delta) / m}
$$

が成立します．よって，$\varepsilon_{\text {gen }}(m, \mathcal{H}, \delta):=2 H \sqrt{\ln (|\mathcal{H}| / \delta) / m}$になります．

---

このような仮定の元，次が成立します．

**定理（BLin-UCBのサンプル効率）**

$\delta \in (0, 1)$とバッチサイズ$m \in \mathbb{N}^+$を固定します．
仮説集合$\mathcal{H}$が次元$d$のBilinear classであり，実現可能であり，さらにBilinear-UCBアルゴリズムが$\mathcal{H}$と$\ell$にアクセスできるとします．
パラメータは次を満たすとします：

$$
T=\left\lceil 2 H d \ln \left(4 H d\left(\frac{B_X^2 B_W^2}{\varepsilon_{\text {gen }}^2(m, \mathcal{H}, \delta /(T H))}+1\right)\right)\right\rceil, \quad R=\sqrt{T} \epsilon_{g e n}(m, \mathcal{H}, \delta /(T H))
$$

$\pi$をBilinear-UCBから返ってっきた方策とします．このとき，確率$1-\delta$以上で，

$$
V_0^{\star}\left(s_0\right)-V_0^\pi\left(s_0\right) \leq 5 \varepsilon_{\text {gen }} \sqrt{d H^3 \ln \left(4 H d\left(\frac{B_X^2 B_W^2}{\varepsilon_{\text {gen }}^2}+1\right)\right)}
$$

が成立します．ここで使用するサンプル数は合計で$mTH$です．

**証明**

簡単のために$\varepsilon_{\text {gen }}:=\varepsilon_{\text {gen }}(m, \mathcal{H}, \delta / T H)$と書きます．
まず，次が成立します．

---

任意の$h \in[0, \ldots, H-1]$ and $t \in[0, \ldots, T-1]$について，確率$1-\delta$以上で，

$$
\sup _{g \in \mathcal{H}}\left|\mathbb{E}_{\mathcal{D}_{t, h}}\left[\ell_{f_t}\left(s, a, s^{\prime}, g\right)\right]-\mathbb{E}_{s \sim d_h^{\pi_f}, a \sim \pi_{e s t}(s), s^{\prime} \sim P_h(\cdot \mid s, a)}\left[\ell_{f_t}\left(s, a, s^{\prime}, g\right)\right]\right| \leq \varepsilon_{g e n}
$$

が成立する．

**証明**

これは$D_{i, h}$が$s \sim d_h^{\pi_{f_t}}, a \sim \pi_{e s t}(s), s^{\prime} \sim P_h(\cdot \mid s, a)$からi.i.d.にサンプルされることと，上で定義した仮定から明らかに成立します．

---

**Optimism**

このイベントが成立するとしましょう．そして$R:=\sqrt{T} \varepsilon_{g e n}$とします．
このとき，任意の$t\in [T]$について，$f^\star$はアルゴリズム内の制約付き最適化問題の実行可能解です．
さらに，$V_{0, f_t}\left(s_0\right) \geq V_0^{\star}\left(s_0\right)$, for all $t \in[T]$が成立します．

**証明**

Bilinear Classの定義から，

$$
\left|\mathbb{E}_{s \sim d_h^{\pi_{f t}}, a \sim \pi_{e s t}(s), s^{\prime} \sim P_h(\cdot \mid s, a)}\left[\ell_{f_t}\left(s, a, s^{\prime}, f^{\star}\right)\right]\right|=\left|\left\langle W_h\left(f^{\star}\right)-W_h\left(f^{\star}\right), X_h\left(f_t\right)\right\rangle\right|=0, \forall t
$$

が成立します．
さらに上のイベントから，
$$
\begin{aligned}
\left|\mathbb{E}_{\mathcal{D}_{t, h}}\left[\ell_{f_t}\left(s, a, s^{\prime}, f^{\star}\right)\right]\right| & =\left|\mathbb{E}_{\mathcal{D}_{t, h}}\left[\ell_{f_t}\left(s, a, s^{\prime}, f^{\star}\right)\right]-\mathbb{E}_{s \sim d_h^{\pi_t}, a \sim \pi_{e s t}(s), s^{\prime} \sim P_h(\cdot \mid s, a)}\left[\ell_{f_t}\left(s, a, s^{\prime}, f^{\star}\right)\right]\right| \\
& \leq \sup _{g \in \mathcal{H}}\left|\mathbb{E}_{\mathcal{D}_{t, h}}\left[\ell_{f_t}\left(s, a, s^{\prime}, g\right)\right]-\mathbb{E}_{s \sim d_h^{\pi_f}, a \sim \pi_{e s t}(s), s^{\prime} \sim P_h(\cdot \mid s, a)}\left[\ell_{f_t}\left(s, a, s^{\prime}, g\right)\right]\right| \\
& \leq \varepsilon_{g e n}
\end{aligned}
$$
も成り立ちます．よって，

$$
\sum_{i=0}^{t-1}\left(\mathbb{E}_{\mathcal{D}_{i, h}} \ell_{f_i}\left(s, a, s^{\prime}, f^{\star}\right)\right)^2 \leq T \varepsilon_{g e n}^2
$$

です．
よって補題の１つ目は成立します．

補題の２つ目は１つ目から直ちに成立します．

---

**エピソードごとのリグレット**

任意の$t \in [0, \dots, T-1]$について，次が成立します．

$$
V_0^{\star}\left(s_0\right)-V_0^{\pi_{f_t}}\left(s_0\right) \leq \sum_{h=0}^{H-1}\left|\left\langle W_h\left(f_t\right)-W_h\left(f^{\star}\right), X_h\left(f_t\right)\right\rangle\right|
$$

証明はOptimismからすぐにわかるので省略します．

---

最後に全部合体してRegretを証明します．Elliptical potential lemmaを使います．

$\Sigma_{t, h}:=\lambda I+\sum_{\tau=0}^{t-1} X_h\left(f_\tau\right) X_h^{\top}\left(f_\tau\right)$とします（$\lambda$は後で設定します）

まず，任意の$h$について次を満たすような$t$が存在することを示します．

$$
\left\|X_h\left(f_t\right)\right\|_{\Sigma_{t, h}^{-1}}^2 \leq \exp \left(\frac{H d}{T} \ln \left(1+\frac{T B_X^2}{d \lambda}\right)\right)-1, \quad\left\|W_h\left(f_t\right)-W_h\left(f^{\star}\right)\right\|_{\Sigma_{t, h}}^2 \leq 4 \lambda B_W^2+4 T \varepsilon_{g e n}^2
$$

Linear Banditでやったようにすれば，

$$
\sum_{t=0}^{T-1} \ln \left(1+\left\|X_h\left(f_t\right)\right\|_{\Sigma_{t, h}^{-1}}^2\right) \leq \ln \frac{\operatorname{det}\left(\Sigma_{T, h}\right)}{\operatorname{det}(\lambda I)} \leq d \ln \left(1+\frac{T B_X^2}{d \lambda}\right)
$$

が成り立ちます．変形して，

$$
\frac{1}{T} \sum_{t=0}^{T-1} \sum_{h=0}^{H-1} \ln \left(1+\left\|X_h\left(f_t\right)\right\|_{\Sigma_{t, h}^{-1}}^2\right) \leq \frac{H d}{T} \ln \left(1+\frac{T B_X^2}{d \lambda}\right)
$$

が成り立ちます．平均値の定理から，$\sum_{h=0}^{H-1} \ln \left(1+\left\|X_h\left(f_t\right)\right\|_{\Sigma_{t, h}^{-1}}^2\right) \leq \frac{H d}{T} \ln \left(1+T B_X^2 /(d \lambda)\right)$であるような$t$が存在します．よって，
$\ln \left(1+\left\|X_h\left(f_t\right)\right\|_{\Sigma_{t, h}^{-1}}^2\right) \leq \frac{H d}{T} \ln \left(1+T B_X^2 /(d \lambda)\right)$が任意の$h$について成立します．

続いて，$\pi_{e s t}$の定義と$f_t$が実行可能解であることから，

$$
\begin{aligned}
& \sum_{i=0}^{t-1}\left(\mathbb{E}_{s \sim d_h^{\pi_f}, a \sim \pi_{e s t}(s), s^{\prime} \sim P_h(\cdot \mid s, a)}\left[\ell_{f_i}\left(s, a, s^{\prime}, f_t\right)\right]\right)^2 \leq 2 \sum_{i=0}^{t-1}\left(\mathbb{E}_{\mathcal{D}_{i, h}}\left[\ell_{f_i}\left(s, a, s^{\prime}, f_t\right)\right]\right)^2 \\
& \quad+2 \sum_{i=0}^{t-1}\left(\mathbb{E}_{s \sim d_h^{\pi_f}, a \sim \pi_{e s t}(s), s^{\prime} \sim P_h(\cdot \mid s, a)}\left[\ell_{f_i}\left(s, a, s^{\prime}, f_t\right)\right]-\mathbb{E}_{\mathcal{D}_{i, h}}\left[\ell_{f_i}\left(s, a, s^{\prime}, f_t\right)\right]\right)^2 \leq 2 T \varepsilon_{g e n}^2+2 T \varepsilon_{\text {gen }}^2=4 T \varepsilon_{\text {gen }}^2
\end{aligned}
$$

が成り立ちます．Bilinear classの定義から，

$$
\begin{aligned}
& \left\|W_h\left(f_t\right)-W_h\left(f^{\star}\right)\right\|_{\Sigma_{t, h}}^2=\sum_{i=0}^{t-1}\left(\left(W_h\left(f_t\right)-W_h\left(f^{\star}\right)\right)^{\top} X_h\left(f_i\right)\right)^2+\lambda\left\|W_h\left(f_t\right)-W_h\left(f^{\star}\right)\right\|_2^2 \\
& =\sum_{i=0}^{t-1}\left(\mathbb{E}_{s \sim d_h^{\pi_f}, a \sim \pi_{e s t}(s), s^{\prime} \sim P_h(\cdot \mid s, a)}\left[\ell_{f_i}\left(s, a, s^{\prime}, f_t\right)\right]\right)^2+\lambda\left\|W_h\left(f_t\right)-W_h\left(f^{\star}\right)\right\|_2^2 \leq 4 T \varepsilon_{\text {gen }}^2+4 \lambda B_W^2,
\end{aligned}
$$

です．
さて，上の$t$について考えましょう．このとき，

$$
\begin{aligned}
V_0^{\star}\left(s_0\right)-V_0^{\pi_{f_t}}\left(s_0\right) & \leq \sum_{h=0}^{H-1}\left|\left(W_h\left(f_t\right)-W_h\left(f^{\star}\right)\right)^{\top} X_h\left(f_t\right)\right| \\
& \leq \sum_{h=0}^{H-1}\left\|W_h\left(f_t\right)-W_h\left(f^{\star}\right)\right\|_{\Sigma_{t, h}}\left\|X_h\left(f_t\right)\right\|_{\Sigma_{t, h}^{-1}} \\
& \leq H \sqrt{4 \lambda B_W^2+4 T \varepsilon_{g e n}^2} \sqrt{\exp \left(\frac{H d}{T} \ln \left(1+\frac{T B_X^2}{d \lambda}\right)\right)-1}
\end{aligned}
$$

が成立します．

最後に，$1 / \lambda=B_W^2 / \varepsilon_{g e n}^2+1 / B_X^2$としましょう．
$T=\left\lceil 2 H d \ln \left(4 H d\left(\frac{B_X^2 B_W^2}{\varepsilon_{g e n}^2}+1\right)\right)\right\rceil$とすると，

$$
\frac{H d}{T} \ln \left(1+\frac{T B_X^2}{d \lambda}\right) \leq \frac{H d}{T} \ln \left(1+\frac{T}{d} \cdot\left(\frac{B_X^2 B_W^2}{\varepsilon_{g e n}^2}+1\right)\right) \leq \frac{H d}{T} \ln \left(2 T\left(\frac{B_X^2 B_W^2}{\varepsilon_{g e n}^2}+1\right)\right) \leq 1
$$

です．最後の部分では，If $T=2 a \ln (2 a b)$ then $a \ln (b T) / T \leq 1$であることを使いました．

よって，

$$
\begin{aligned}
V_0^{\star}\left(s_0\right)-V_0^{\pi}\left(s_0\right) & \leq H \sqrt{2\left(4 \lambda B_W^2+4 T \varepsilon_{\text {gen }}^2\right)} \\
& \leq H \sqrt{8 \varepsilon_{\text {gen }}^2+16 \varepsilon_{\text {gen }}^2 H d \ln \left(4 H d\left(\frac{B_X^2 B_W^2}{\varepsilon_{\text {gen }}^2}+1\right)\right)} \\
& \leq 5 \varepsilon_{\text {gen }} \sqrt{d H^3 \ln \left(4 H d\left(\frac{B_X^2 B_W^2}{\varepsilon_{\text {gen }}^2}+1\right)\right)} .
\end{aligned}
$$

が成立します．
