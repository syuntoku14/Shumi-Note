# 位相線形空間

![MATh_FA_basic.ipynb](MATh_FA_basic.ipynb)の続き

前回は関数解析の文脈で無限次元の線形空間を考えたかったわけですね．
フーリエ級数展開のように，線形代数に無限次元の和（収束）を考えたいために位相を導入しました．
位相としてはどのようなものが良いでしょうか？

前回は開集合を使いましたが，これは抽象度が高すぎてよくわかりません．
そこで，関数のノルムを考えます．


## ノルム（線形）空間

$X$を$\mathbb{R}$体上の線形空間とする．
次を満たす$\|\cdot\|: X \rightarrow \mathbb{R}$を$X$上のノルムといいます．

任意の$a, b\in X$および$\lambda\in \mathbb{R}$に対して，

1. 正定値性：$\|a\| \geq 0$で$a=0 \Leftrightarrow\|a\|=0$
2. 同次性・斉次性：$\|\lambda a\|=|\lambda|\|a\|$
3. 劣加法性：$\|\dot{a}+b\| \leq\|a\|+\|b\|$

$(X, \|\|)$をノルム（線形）空間と呼びます．単に$X$をノルム（線形）空間と呼ぶこともあります．

**関数のノルムと距離の例**

* $C^0(X)=\{f: X \rightarrow \mathbb{R} \mid f=$ 連続関数 $\}$：$X$上の連続関数の集合は線形空間です
* $L^1(X):=\left\{f: X \rightarrow \mathbb{R}\left|\int_X\right| f(x) \mid d x<\infty\right\}$：$X$上の可積分関数の集合は線形空間です

また，それぞれ次のようなノルムが取れます：

$$
\|f\|_{C^0(X)}=\|f\|_{\infty}=\sup _{x \in X}|f(x)|, \quad\|f\|_{L^1(X)}=\int_X|f(x)| d x
$$

これを使うと，ノルムから定まる距離$\|f - g\|$が考えられます：

$$
\sup _{x \in X}|f(x)-g(x)| . \quad \int_X|f(x)-g(x)| d x
$$

**半ノルム**

劣加法性と同次性のみ成り立つ場合を半ノルムといいます．


### 集合の有界性

* $(X, d)$を距離空間とします．$K$を$X$の部分集合とします．集合$K$は，次を満たすときに有界であるといいます：

$$
\exists M>0, \exists a \in K, \forall x \in K: \mathrm{d}(x, a) \leq M
$$
さらに次が成り立ちます：
$$
\Leftrightarrow \exists M>0, \forall a \in K, \forall x \in K: \mathrm{d}(x, a) \leq M
$$

証明： $\Leftarrow$は明らかです．$\Rightarrow$は三角不等式を使うと$\forall x \in K$，$\forall h \in K$で$d(x, b) \leq d(x, a)+d(b, a) \leq 2 M$となる$a \in K$が存在することがわかるので，成り立ちます．

補足：この２つは使用用途によって使い分けます．いろいろな場合で使い分けられると便利なので，同値な定義をいくつか導入するのは良い行いです．

### 距離空間におけるコンパクト性

以下ではコンパクト性を導入します．コンパクト集合$K$上の連続関数$f$は，
* 最大値・最小値を持つので，最適化問題として定式化できる
* 一様連続 → $K$上での近さで$f$の近さを議論できる
* その集合を$\mathcal{C}(K)$とすると，一様（無限大）ノルム$\|\cdot\|_\infty$に対して距離空間$\left(\mathcal{C}(K),\|\cdot\|_{\infty}\right)$が完備です．

ここで，完備距離空間とは，空間内の任意のコーシー点列がその空間内に極限を持つ空間のことです．関数の劣$f_n$がコーシー列になっていれば，その極限$lim f_n$がその空間の中にあることを保証できます．これを使うと$f_n (n \to \infty)$の一様収束が言えます．

距離空間においては点列コンパクト性とコンパクト性は同値です．**しかし，一般の位相空間では異なります．**

---

**点列コンパクト**

距離空間$(X, d)$において，$X$の部分集合$K$が点列コンパクトであるとは，$K$に含まれる任意の点列$\left(x_n\right)_{n \in \mathbb{N}}$に対して，$K$の中の点に収束する部分劣が存在すること．

---

位相空間については次のコンパクト性が考えられます．

**開被覆によるコンパクト性の定義**

位相空間$X$がコンパクト＝任意の開被覆$\mathcal{U}$に対して**有限個の要素** $U_1, \ldots, U_n \in \mathcal{U}$ で $X=\bigcup_{i=1}^n U_i$とできること．

ここで$X$の部分集合の集まり$\left\{U_\lambda\right\}_{\lambda \in \Lambda}$が開被覆であるとは，すべての$U_\lambda$が開集合であり，$X \subset \bigcup_{\lambda \in \Lambda} U_\lambda$であること．

位相空間に対してコンパクト性があるとはつまり，「その空間が有限な大きさである」ことを表してます．
特に有限次元空間$\mathbb{R}^n$では，部分空間がコンパクトであることと有界閉集合であることは同値です（Heine-Borelの定理）．

無限次元空間（関数の空間など）では，部分空間がコンパクト$\Rightarrow$有界閉集合は成り立ちますが，逆は成り立ちません．


## 線形作用素

$X, Y$を$\mathbb{R}$体上のノルム線形空間とします．写像$T: X \to Y$が線形作用素であるとは，
$$
T(\alpha x+y)=\alpha T x+T y, \quad x, y \in X, \alpha \in \mathbb{R}
$$
を満たすこと．

$X$から$Y$への線形作用素の全体を$\mathcal{L}(X, Y)$と表し，$T_1, T_2 \in L(X, Y), x \in X, \alpha, \beta \in \mathbb{R}$に対して
$$
\left(\alpha T_1+\beta T_2\right)(x)=\alpha T_1 x+\beta T_2 x
$$
と定義すると，$\mathcal{L}(X, Y)$は線形空間になります．

* 連続性：$T \in L(X, Y)$が$x_0 \in X$で連続であるとは，
$$
\left\|x-x_0\right\| \rightarrow 0 \Rightarrow\left\|T x-T x_0\right\| \rightarrow 0
$$
を満たすとき．特に$X$の任意の点$x_0 \in X$で連続のとき，$X$上で連続であるという．
* 有界性：また，$T \in L(X, Y)$が$M \in \mathbb{R}_{> 0}$に対して$\|T x\| \leq M\|x\| \quad x \in X$を満たすとき，$T$は有界であるといいます．

実は連続性と有界性は同値です．
$T\in L(X, Y)$に対して，
1. $X$上で連続
2. $0 \in X$で連続
3. 有界

は同値になります．

**証明**

1→2は明らかです．

2→3を示しましょう．
$0 \in X$で連続であることは，

$$
\Leftrightarrow{ }^{\forall} \varepsilon>0,{ }^{\exists} \delta>0,\|x\|<\delta \Rightarrow\|T x\|<\varepsilon .
$$

です．ここで，
$$
{ }^{\forall} x \neq 0 \in X, \quad\left\|\frac{\delta}{2\|x\|} x\right\|<\delta
$$
なので，
$$
\|T x\|=\frac{2\|x\|}{\delta}\left\|T\left(\frac{\delta}{2\|x\|} x\right)\right\|<\frac{2 \varepsilon}{\delta}\|x\|
$$
が成り立ちます．

3→1を示しましょう．$\|T x-T y\|=\|T(x-y)\| \leq M\|x-y\|$なので，$\|x - y \|\to 0$のとき，$\|Tx - Ty\|\to 0$です．


### 線形汎関数と劣線型性

線形空間からその係数体への線形写像を線形汎関数といいます．つまり，$\mathbb{R}$体上の線形空間$X$上の線形汎関数$\ell$とは$\ell: X \to \mathbb{R}$である線形写像です．（$X$から体へ写像する関数です）

$\mathbb{R}$体上の線形空間$X$上の汎関数$\ell: X\to \mathbb{R}$に対して，
$$
\begin{aligned}
\ell(x+y) & \leq \ell(x)+\ell(y)(x, y \in X) \\
\ell(\lambda x) & =\lambda \ell(x)\left(x \in X, \lambda \in \mathbb{R}_{\geq 0}\right)
\end{aligned}
$$
を満たす写像を劣線形汎関数といいます．


### ゲージ汎関数（ミンコフスキー汎関数）

実数$\mathbb{R}$体上のノルム線形空間$X$の部分空間$K \subseteq X$を凸集合として，$0$を内点として含むとします．

ゲージ汎関数$p_K: X \to \mathbb{R}_{\geq 0}$とは，
$$
p_K(x)=\inf \left\{a \in \mathbb{R}_{>0} \left\lvert\, \frac{1}{a} x \in K\right.\right\}
$$
で定義される関数です．
つまり，$x$に$1/a$しても$K$に入ってるような$a$のうち一番小さいものを出します．$1$が必ず入ってることはわかるので，$1$以下の値です．

ゲージ汎関数は半ノルムです．また，ゲージ汎関数を使うと$x$が$K^\circ$の内点であるかを判別できます．
次が成立します：

$$
p_K(x)<1 \Leftrightarrow x \in K^{\circ}
$$

$\Rightarrow$の証明：
$$
\begin{aligned}
& p_K(x)<1 \Rightarrow^{\exists} \alpha \in(0,1), \frac{1}{\alpha} x \in K . \\
& 0 \in K^{\circ} \Leftrightarrow{ }^{\exists} \epsilon>0, U_\epsilon(0) \subseteq K
\end{aligned}
$$
ここで$K$は凸集合なので，$\forall y \in U_\epsilon(0), \alpha \frac{1}{\alpha} x+(1-\alpha) y \in K$です．ここで，$z=x + (1-\alpha) y \in K$とおくと，
$$
\|z-x\|=\|(1-\alpha) y\|<(1-\alpha) \epsilon
$$
なので，
$$
{ }^{\exists} \alpha \in(0,1),{ }^{\exists} \epsilon>0,{ }^{\forall} z \in U_{(1-\alpha) \epsilon}(x) \Rightarrow z \in K
$$
であり，$U_{(1-\alpha) \epsilon}(x) \subseteq K$が成り立つので，$x \in K^\circ$が成り立ちます．


実はこの証明は$K$が開集合のときはめっちゃ簡単です．

$$
p_K(x)<1 \Rightarrow \exists \alpha \in(0,1), \frac{1}{\alpha} x \in K
$$
であり，$K$は凸集合で$0$を内点として含むので，
$$
\alpha \frac{1}{\alpha} x+(1-\alpha) 0 \in K .
$$
が成り立ちます．（$K$が開集合のとき，$K = K^\circ$なのですぐにわかります）

---

**別解**

$x \notin K^{\circ}$を仮定すると，
$$
\exists y \in X, \forall \epsilon>0, \exists|t|<\epsilon, x+t y \notin K .
$$
であり，$p_K(x+t y)>1$を満たす$y$と$t$が存在します．劣加法性より
$$
1<p_K(x+t y) \leq p_K(x)+p_K(t y)
$$
なので，$p_K(x)>1-p_K(t y)$が成り立ちます．最後に$-|t|>-\epsilon$と同次性から，
$$
p_K(x)>1-|t| p_K(y)>1-\epsilon p_K(y)
$$
より，よって$p_K(x) \geq 1$です．

---

$\Leftarrow$の証明：
$x \in K^\circ$を仮定すると，$\forall y \in X, \exists \epsilon>0, \forall|t|<\epsilon, x+t y \in K$より，$\forall y \in X, p_K(x+t y) \leq 1$です．$y=x, t>0$とすれば，同次性から$p_K((1+t) x)=(1+t) p_K(x) \leq 1$なので，$p_K(x) \leq 1 /(1+t)<1$が成り立ちます．


---

**別解**

$x \in K^{\circ}$と仮定すると，
$$
{ }^{\exists} \epsilon>0, U_\epsilon(x) \subseteq K, U_\epsilon(x)=\{y \in K:\|x-y\|<\epsilon\}
$$
より，$\forall y \in U_\epsilon(x) \Rightarrow y \in K$です．$y=\left(1+\frac{\epsilon}{2\|x\|}\right) x$とすると，$y \in U_\epsilon(x) \subseteq K$です．つまり，少なくとも$p_K(y) \leq 1$です．同次性より，
$$
\begin{aligned}
p_K(x) & =p_K\left(\left(1+\frac{\epsilon}{2\|x\|}\right)^{-1} y\right)=\left(1+\frac{\epsilon}{2\|x\|}\right)^{-1} p_K(y) \\
& \leq\left(1+\frac{\epsilon}{2\|x\|}\right)^{-1}<1 .
\end{aligned}
$$
です．


## ミンコフスキー和/差

$A \pm B=\{a \pm b \mid a \in A, b \in B\}$のことをミンコフスキー和/差といいます．
$A$と$B$が凸集合ならば，ミンコフスキー和/差も凸集合です．

---

$X$を$\mathbb{R}$体上のノルム線形空間とします．
$A \subseteq X, B \subseteq X$を$X$の凸な空でない部分集合とします．$A$をコンパクト集合，$B$を閉集合とします．このとき，ミンコフスキー和$A+B$および$A-B$は閉集合です．
