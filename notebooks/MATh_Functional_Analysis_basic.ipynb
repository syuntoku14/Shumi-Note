# 関数解析の基礎（位相と開集合，連続関数）

よくある線形代数の講義では有限次元のベクトル空間についての議論を展開します．一方で，物理学における微分方程式や量子力学を考える，
1. 解として関数を扱う
2. 波動関数で物理状態を表現する

など，無限次元の線形空間として関数の空間を考えることがあります．
これが関数解析をしたいモチベーションです（多分）

無限次元は有限次元と同様に線形作用素を分析するのが難しいです．
そこで，「位相的な考察」を導入し，「収束の概念」を加えた無限次元の線形代数を扱えるようにしたのが関数解析です．

まずは位相から学びます．

## 線形空間（ベクトル空間）の復習

集合$X$が体$\mathbb{R}$上の線形空間であるとは，$X$上に以下を満たす演算$+$と$\mathbb{R}$倍が定義されている集合のことです．

1. $(x+y)+z=x+(y+z),{ }^{\forall} x, y, z \in X$
2. $x+y=y+x,{ }^{\forall} x, y \in X$
3. $\mathbf{0} \in X,{ }^{\forall} x \in X, x+\mathbf{0}=\mathbf{0}+x=x$
4. ${ }^{\forall} x \in X,{ }^{\exists} x^{\prime} \in X, x+x^{\prime}=x^{\prime}+x=\mathbf{0}$
5. $(a+b)x = ax + bx$
6. $a(x+y) = ax + by$
7. $(ab)x = a(bx)$
8. $1a = a$

---

**線形部分空間**

$\mathbb{R}$体上の線形空間$X$の空でない部分集合$ S \subseteq X$が線形部分空間とは，任意の$x, y \in S$および$a \in \mathbb{R}$に対して，元の線形空間$X$で定義された和やスカラー席：
$$
x + y \in S , \quad ax \in S
$$
を満たすことをいいます．

線形空間$X$自身や$X$の零元だけからなる集合は$X$の部分空間であり，自明な部分空間と呼ばれます．

---

**線形空間の基底**

体$\mathbb{K}$（実数全体$\mathbb{R}$や複素数全体$\mathbb{C}$など）上の線形空間$V$の$n$個のベクトル集合$B=\{v_1, \cdots, v_n\}$が基底であるとは，

* 線形独立性：
$$
\forall a_1, \cdots, a_n \in \mathbb{K}, \sum_{i=1}^n a_i v_i=0 \Rightarrow a_1=\cdots=a_n=0
$$
* 全域性：
$$
\forall x \in V, \exists a_1, \cdots, a_n \in \mathbb{K}, x_0=\sum_{i=1}^n a_i v_i
$$
の２つを満たすことを言います．また，このような$n$が存在するとき，有限次元であるといいます．

---

さて，ここまでは普通の線形代数の講義で習う内容です．無限次元へ拡張するために，線形空間の基底を一般化しましょう．

体$\mathbb{K}$（実数全体$\mathbb{R}$や複素数全体$\mathbb{C}$など）上の線形空間$V$の部分集合$B \subset V$が基底であるとは，

* 任意の有限部分集合$B_0 \subset B$が線形独立性を持つ
* $\forall x \in V, \exists n \in \mathbb{N}, \exists v_i \in B, \exists a_1, \cdots, a_n \in \mathbb{K}, x=\sum_{i=1}^n a_i v_i$

を満たすことをいいます．
ここで，$B$は無限集合でもかまいません．

関数解析を扱う場合はこれで十分でしょうか？実はまだ不十分です．
たとえばフーリエ変換など，無限個の基底を使って表現したい場合：
$$
x = \sum_{i=1}^\infty a_i v_i
$$
が扱えません．これを扱いたいために位相が必要になります（つまり，$\infty$に対する収束の概念を扱いたいのです）．

## 位相空間

位相空間とは，ある集合に「位相」という構造を定義したものです．
具体的には，集合上に
$$点の近傍,開集合, 閉集合, 閉包など$$
の概念を定義するのに必要十分な構造によって定義されます．
位相は「近さ」の概念を抽象的にしたもの，とよく言われます（後で詳しく見ます）．

このどれか一つの位相構造を定義できれば，他も定義できます（TODO: 証明）．
後で見るように，開集合を使うことが基本です．

（点列の収束を位相構造を作れるため，一般には点列の収束を位相構造とすることはできない．しかし，点列を一般化した有向点族の収束ならば定義できる．なので今回は点列の収束を定義に含めない）

---

**位相空間の定義と開集合**

集合$X$の部分集合系$\mathcal{O}$が以下の条件を満たすとします．例えば実数については部分集合系は実数区間の集まりです．（集合系は集合族とほぼ定義が同じです．集合そのものに興味がある場合は集合系という単語が使われることが慣用的に使われます．例えば$\sigma$加法族は要素になにか操作をすることを前提にしてるので，集合族と呼ばれることが多いです．）：

1. $\varnothing, X \in \mathcal{O}$
2. ${ }^{\forall} U_1, U_2 \in \mathcal{O}, U_1 \cap U_2 \in \mathcal{O}$
   1. つまり，$\mathcal{O}$に属する有限個の集合の共通部分は$\mathcal{O}$に属します．
3. 任意の添字集合$\Lambda$に対して，$\bigcup_{\lambda \in \Lambda} U_\lambda \in \mathcal{O}$

このとき，$\mathcal{O}$を$X$の位相といい，$(X, \mathcal{O})$を位相空間という．
（単に$X$を位相空間ということもある．）

$\mathcal{O}$の元を開集合という．また，位相を開集合系ともいう．上の$1~3$を開集合の定義ということもある．

つまり，位相空間は開集合の集合です．

---

**閉集合**

$(X, \mathcal{O})$を位相空間とする．
開集合の補集合を閉集合という．この位相空間の閉集合をすべて集めた集合を閉集合系という：

$$
\mathcal{F}=\left\{U^c \mid U \in \mathcal{O}\right\}
$$

閉集合系は以下を満たします：
1. $\varnothing, X \in \mathcal{F}$
2. $\forall F_1, F_2 \in \mathcal{F}, F_1 \cup F_2 \in \mathcal{F}$
3. 任意の添字集合$\Lambda$に対して，$\bigcap_{\lambda \in \Lambda} F_\lambda \in \mathcal{F}$

逆に閉集合を先に上の３つで定義して，その補修号として開集合を定義することもあります．

---

**連続関数（写像）の定義**

関数解析の文脈では連続関数の話がしばしば出てきます．後で位相を使った解析について見るので，連続関数の定義を導入しましょう．

位相空間$(X, \mathcal{O}),\left(X^{\prime}, \mathcal{O}^{\prime}\right)$の間の写像$f: X \rightarrow X^{\prime}$が連続であるとは，

$$
U^{\prime} \in \mathcal{O}^{\prime} \Longrightarrow f^{-1}\left(U^{\prime}\right) \in \mathcal{O}
$$

が成り立つことを言います．すなわち，任意の$X'$の開集合の逆像が$X$の開集合になることを言います．

---

**補足：逆像の定義**

ここで，逆像は逆写像とは全く違うことに注意しましょう．[参考](https://seetheworld1992.hatenablog.com/entry/2020/10/29/063434)．逆写像は全単射のときにしか存在しません．逆像は常に存在します．

$f: X\to Y$とします．$A \subset X$に対して，
$$
f(A)=\{f(x) \in Y \mid x \in A\}
$$
を$f$による$A$の像といいます．$X$全体の像を$\operatorname{Im} f \equiv f(X)$と書いたりもします．また，$B\subset Y$に対して，
$$
f^{-1}(B)=\{x \in X \mid f(x) \in B\}
$$
を$f$による$B$の逆像といいます．

---


なんだかよくわからないですね．一旦実数空間上の話を確認してみましょう．


**$\mathbb{R}$上の連続関数（写像）の定義**

特に実数空間上では，連続関数はより直感的に定義されます．

関数$f: \mathbb{R}\to \mathbb{R}$が点$x_0 \in \mathbb{R}$で連続であるとは，

$$
\begin{aligned}
&{ }^{\forall} \epsilon>0,{ }^{\exists} \delta>0, \\
& \\
&{ }^{\forall} x \in \mathbb{R},\left|x-x_0\right|<\delta \Rightarrow\left|f(x)-f\left(x_0\right)\right|<\epsilon
\end{aligned}
$$

が成り立つことを言います．
任意の$x_0 \in \mathbb{R}$において$f$が連続であるとき，$f$は連続であるといいます．

$\epsilon$-$\delta$論法ってやつですね．

これについて，
この定義を開集合で書き換えてみましょう（多分以下の話は必要十分条件になってる？）．
$a$の$\epsilon$近傍を
$$
U_{\varepsilon}(a)=\{x \in \mathbb{R}|| x-a \mid<\varepsilon\}
$$
とすると，上の定義から，連続な関数は

$$
\begin{aligned}
{ }^{\forall} \epsilon & >0,{ }^{\exists} \delta>0, \\
{ }^{\forall} x & \in \mathbb{R}, x \in U_\delta\left(x_0\right) \Rightarrow f(x) \in U_\epsilon\left(f\left(x_0\right)\right)
\end{aligned}
$$
を満たします．更に書き換えると，

$$
{ }^{\forall} \epsilon>0,{ }^{\exists} \delta>0, f\left(U_\delta\left(x_0\right)\right) \subseteq U_\epsilon\left(f\left(x_0\right)\right)
$$
を連続関数$f$は満たします．
**つまり，$x_0$の近傍を$f$で写像すると，$f(x_0)$の近傍に必ず含まれます．近傍→近傍の関係が成り立ってるわけですね**


これに対して逆像を取ってみましょう．
$$
{ }^{\forall} \epsilon>0,{ }^{\exists} \delta>0, U_\delta\left(x_0\right)\subseteq f^{-1}\left(U_\epsilon\left(f\left(x_0\right)\right)\right)
$$

実はこの$f^{-1}\left(U_\epsilon\left(f\left(x_0\right)\right)\right)$は開集合になっています．

**開集合であることの補足**

実数空間については，$O \subseteq \mathbb{R}^n$が開集合であることは次と同値です：

$$
{ }^{\forall} x \in O,{ }^{\exists} \epsilon>0 \text { s.t } U_\epsilon(x) \subseteq O
$$
つまり，任意の点$x\in O$で，半径$\epsilon$の開球が全てその集合$O$に含まれてる状況ですね．
この定義から，上の
$f^{-1}\left(U_\epsilon\left(f\left(x_0\right)\right)\right)$が開集合であることはすぐにわかります．

---

上の変換は，**実数空間における連続の概念を開集合を使って説明しました**．
続いて距離の概念について説明します．

## 距離空間

後で再び位相の話に戻りますが，ここで一旦距離の話をはさみます．
位相とは「近さ」の概念を抽象的にしたもの，という話をしました．距離を定義すると，この「近さ」が分かりやすくなります（位相は集合系で定義されてるので，距離関数を使った定義より抽象的です．後で再び集合系による定義の話に戻ります）．

$X$を集合とする．２変数関数$d: X \times X \to \mathbb{R}$が任意の$a, b, c \in X$に対して

1. 正定値性：$d(a, b) \geq 0$で, $a=b \Leftrightarrow d(a, b)=0$
2. 対称性：$d(a, b)=d(b, a)$
3. 三角不等式：$d(a, b) \leq d(a, c)+d(c, a)$

を満たすとき，$d$は距離（関数）と呼ばれ，$(X, d)$を距離空間と呼びます．

---

**距離空間における開集合と閉集合**

$(X, d)$を距離空間とする．このとき，
$$
U_{\varepsilon}(a)=\{x \in X \mid d(x, a)<\varepsilon\}
$$
を中心$a \in X$かつ半径が$\varepsilon$の開球と呼びます．（不等号が$<$であることに注意しましょう）．

また，この開球を使って開集合を定義できます：
* $A$が開集合$\Longleftrightarrow A^c=X \backslash A$が閉集合
* $A$が開集合$\Leftrightarrow \forall a \in A, \exists \varepsilon>0: U_{\varepsilon}(a) \subseteq A$
* $B$が閉集合$\Longleftrightarrow \forall b \in B^c, \exists \varepsilon>0: U_{\varepsilon}(b) \cap B=\varnothing$

また，開球を使って

$$
\mathcal{O}=\left\{O \subset X \mid{ }^{\forall} a \in O,{ }^{\exists} \varepsilon>0 \text { s.t. } U_{\varepsilon}(a) \subset O\right\}
$$

とすると，$\mathcal{O}$は位相空間の条件を満たすので位相になります．

この定義は結構嬉しいです．開集合の定義は，その元$a \in A$の近傍の性質だけで全体$A$の性質が定義されています．近傍を見るだけで全体がわかる，この性質を後でめっちゃ使います．

ここで見たように，距離空間における開集合は距離関数による開球で自然に定義されます．位相は開集合の集合なので，開球で位相は自然に解釈できます．


## 集合系と近さ

一番最初に定義した位相は集合系による定義でした．これに「近さ」の概念を見出してみます．

例えば集合$S=\{1, 2, 3, 4, 6\}$について，

$$
\begin{aligned}
& \mathcal{S}_1=\{\varnothing,\{1,2,4\},\{3,6\}, S\} \\
& \mathcal{S}_2=\{\varnothing,\{1,3\},\{2,4,6\}, S\} \\
& \mathcal{S}_3=\{\varnothing,\{6\},\{3,6\},\{4,6\},\{3,4,6\}, S\}
\end{aligned}
$$

なる集合系を考えてみましょう．なんとなくですが，例えば$S_1$をみると，$1, 2, 4$はなんとなく近そうです．一つ一つの集合が「カテゴリ」を表現し，なんとなく要素感の近さを表現していそうです．

このように，集合だけ考えても「近さ」を定義することができます．

ここで，改めて上で学んだ連続関数の定義をおさらいしましょう．

---

位相空間$(X, \mathcal{O}),\left(X^{\prime}, \mathcal{O}^{\prime}\right)$の間の写像$f: X \rightarrow X^{\prime}$が連続であるとは，

$$
U^{\prime} \in \mathcal{O}^{\prime} \Longrightarrow f^{-1}\left(U^{\prime}\right) \in \mathcal{O}
$$

が成り立つことを言います．すなわち，任意の$X'$の開集合の逆像が$X$の開集合になることを言います．

さらに実数値連続関数のときは，

$$
\begin{aligned}
& { }^{\forall} x_0 \in \mathbb{R},{ }^{\forall} \epsilon>0,{ }^{\exists} \delta>0, f\left(U_\delta\left(x_0\right)\right) \subseteq U_\epsilon\left(f\left(x_0\right)\right) \\
& { }^{\forall} x_0 \in \mathbb{R},{ }^{\forall} \epsilon>0,{ }^{\exists} \delta>0, U_\delta\left(x_0\right) \subseteq f^{-1}\left(U_\epsilon\left(f\left(x_0\right)\right)\right)
\end{aligned}
$$
で書き換えられます．

---

書き換えの１つ目はかなり直感的でわかりやすいですね．上でやったように，これは近傍→近傍の写像が成立することを述べてます．
なぜわざわざ分かりづらい逆像で定義するのでしょうか？

まず，像と逆像について次が成立します：

写像$f: X \rightarrow Y, \quad A_1, A_2 \subset X, \quad B_1, B_2 \subset Y$について，

$$
\begin{aligned}
& f\left(A_1 \cup A_2\right)=f\left(A_1\right) \cup f\left(A_2\right) \\
& f^{-1}\left(B_1 \cup B_2\right)=f^{-1}\left(B_1\right) \cup f^{-1}\left(B_2\right) \\
& f\left(A_1 \cap A_2\right) \subset f\left(A_1\right) \cap f\left(A_2\right) \\
& f^{-1}\left(B_1 \cap B_2\right)=f^{-1}\left(B_1\right) \cap f^{-1}\left(B_2\right) \\
& f\left(A_1\right) \backslash f\left(A_2\right) \subset f\left(A_1 \backslash A_2\right) \\
& f^{-1}\left(B_1\right) \backslash f^{-1}\left(B_2\right)=f^{-1}\left(B_1 \backslash B_2\right)
\end{aligned}
$$

これを見ると，**逆像のほうが像よりも等号成立が多く，性質が良い**ことがわかります．なので逆像を使うんですね．

ちなみに一般には次が成り立ちます：

$\Lambda$を任意の添字の集合とします．
写像$f: X \rightarrow Y$，$X$の部分集合族$\left\{A_\lambda \mid A_\lambda \subset X, \lambda \in \Lambda\right\}$，$Y$の部分集合族$\left\{B_\lambda \mid B_\lambda \subset Y, \lambda \in \Lambda\right\}$に対して，

$$
\begin{aligned}
& f\left(\bigcup_\lambda A_\lambda\right)=\bigcup_{\lambda \in \Lambda} f\left(A_\lambda\right) \\
& f^{-1}\left(\cup_{\lambda \in \Lambda} B_\lambda\right)=\bigcup_{\lambda \in \Lambda} f^{-1}\left(B_\lambda\right) \\
& f\left(\bigcap_{\lambda \in \Lambda} A_\lambda\right) \subset \bigcap_{\lambda \in \Lambda} f\left(A_\lambda\right) \\
& f^{-1}\left(\bigcap_{\lambda \in \Lambda} B_\lambda\right)=\bigcap_{\lambda \in \Lambda} f^{-1}\left(B_\lambda\right)
\end{aligned}
$$

が成立します．$\Lambda$は可算（無限）である必要はないです．$\mathbb{R}$のような非可算無限集合でも成立します．

## 近傍と開近傍

関数解析ではしばしば**稠密**という単語がでてきます．これを正確に定義するために，近傍と開近傍という概念を導入しましょう．
また，近傍や開近傍を使うと集合の内部が定義され，内部が定義できると微分が定義できます．

$x\in X$の$X$における近傍$V(x) \subseteq X$とは，ある開集合$U \subseteq V(x)$が存在して，

$$x \in U \subseteq V(x)$$

となることを言います．特に$V(x)$が開集合のとき開近傍といいます．
定義より，$X$も近傍であることに注意しましょう．

特に距離空間では，開球
$$U_{\varepsilon}(a)=\{x \in X \mid d(x, a)<\varepsilon\}$$
は$\varepsilon$近傍と呼ばれます．

**位相空間と内点**

位相空間$(X, \mathcal{O})$を考えます．$A$を$X$の部分集合とします．このとき，

* $x$が$A$の内点であるとは，${ }^{\exists} U \in \mathcal{O}, x \in U \subseteq A$ことを言います．
* 特に$X$が距離空間ならば，${ }^{\exists} \epsilon>0, U_\epsilon(x) \subseteq A$であることと同じです．なので距離空間では$\varepsilon$近傍で書けます．
* $X$が実数$\mathbb{R}$体上のノルム線形空間であれば，$\forall y \in X, \exists \epsilon>0,|t|<\epsilon \Rightarrow x+t y \in A$であることと同じです．

**内部とコア**

$X$が実数$\mathbb{R}$体上のノルム線形空間である場合，$x\in X$が$A$の内部の点であるとは，

$$
\forall y \in X, \exists \epsilon>0,|t|<\epsilon \Rightarrow x+t y \in A
$$

のことでした．つまり，$x$から任意の方向へ少し動いてもまだ$A$の中に留まるわけですね．

一方で，$x\in X$が$A$のコアの点とは，

$$
\forall y \in X, \exists \epsilon>0, 0<t<\epsilon \Rightarrow x+t y \in A
$$
であることを言います．つまり，コアでは正の方向にしか行けません．
$x$が$A$の境界上にある場合，内点にはなれませんが，コアになることはできます．内部はコアの部分集合です．


**内部・外部・境界の定義**

$A$を位相空間$X$の部分集合とします．

* $A$の内部：$A$の全ての内点からなる集合．$A^{\circ}$ または$\operatorname{Int} A$と書く．
* $A$の外部：$A$の補集合の内部．$A^{c}$ または$\operatorname{ext} A$と書く．$A^c=\left(A^c\right)^{\circ}$．外部に属する点のことを外点と呼ぶ．
* $A$の境界：$A$の内部にも外部にも属さない点の集合．$\partial A=X \backslash\left(A^{\circ} \cup A^c\right)$と書く．境界に属する点を境界点という．境界点の任意の開近傍は$A$に属する点と属さない点の両方を含む．

**内部と開近傍の和**

$A$を位相空間$X$の部分集合とします．$A=A^\circ$とします（これは$A$が開集合であることの必要十分条件になってます．後で見ます．）

$x \in A^\circ$の$A^\circ$における開近傍を
$V(x) \subseteq A^{\circ}$とすると，

$$
A^{\circ}=\bigcup_{x \in A^{\circ}} V(x)
$$

を満たす開近傍$V$が存在します．

---

**証明**

* $\subseteq$の証明

内部の定義より，任意の$x \in A^\circ$に対して$x \in V(x) \subseteq A$となる開近傍$V(x)$が存在するので，$x \in \bigcup_{x \in A^{\circ}} V(x)$

* $\supseteq$の証明

上の$V$は$V(x)\subseteq A$なので，$A=A^\circ$のときは明らかに成立．

---

続いて次を示しましょう．

$A$を位相空間の部分集合とすると，$A^\circ$は開集合である．

**証明**

内点の定義より，任意の$x \in A^\circ$に対して
$x \in U(x) \subseteq A$となる開集合$U(x)$が存在するので，$A^\circ \subseteq \bigcup_{x \in A^{\circ}} U(x)$

$U$を$A$に含まれる任意の開集合とすると，任意の$x \in U$について$x \in U \subseteq A$が成り立つので，$x$は内点である．よって，
$A$に含まれる任意の開集合$U$について$U \subseteq A^\circ$が成り立つので，$\bigcup_{x \in A^{\circ}} U(x) \subseteq A^{\circ}$である．

よって$A^\circ = \bigcup_{x \in A^\circ}U(x)$なので，開集合の和集合は開集合なので，$A^\circ$は開集合である．

---

さらに次が成立します．

$A$を位相空間の部分集合とすると，$A$が$X$の開集合であることと，$A^\circ = A$であることは同値．

**証明**

$\Rightarrow$の証明

$A$が$X$の開集合ならば，${ }^{\forall} x \in A, x \in A \subseteq A$なので，$x \in A^\circ$．よって$A \subset A^{\circ}$です．内部の定義より$A^{\circ} \subseteq A$なので，$A=A^\circ$


$\Leftarrow$の証明

$A=A^\circ$のとき，$A=A^{\circ}=\bigcup_{x \in A^{\circ}} V(x)$を満たす開集合が存在する．開集合の和は開集合なので，$A$は開集合．


---

**内部と最大の開集合**

$A$を位相空間の部分集合とすると，$A^\circ$は$A$に含まれる最大の開集合である．

**証明**

$U$ を $A$ に含まれる任意の開集合とする．
任意の $x \in U \subseteq A$ に対して，自明に ${ }^\exists U$ s.t. $x \in U \subseteq A$. よって, $x \in A^{\circ}$が成り立つ. すなわち, $A$ の任意の開集合 $U$ に対して $U \subseteq A^{\circ}$ で $A^{\circ}$ は開集合なので, $A^{\circ}$ は最大の開集合である.

ちなみに$A^\circ$が$A$に含まれる最大の開集合であることと，$A^\circ$は$A$に含まれるすべての開集合の和集合であることは同値です．これが上の証明のアイデアです．



## 触点と閉包

内点と内部は開集合に対応していましたが，触点と閉包は閉集合に対応します．

位相空間$(X, \mathcal{O})$を考えます．$A$を$X$の部分集合とします．このとき，

* $x \in X$が$A$の触点であるとは，$x$の任意の開近傍$U(x)$が$A$と交わることである．つまり，
$$
U(x) \cap A \neq \varnothing .
$$
* $A$の触点の集合を$A$の閉包と呼び，$\overline{A}$または$\operatorname{cl} A$と表す
* 触点は, $A$ 自体に含まれているか $A$ の境界に存在する可能性がある. すなわち, $\overline{A}=A^{\circ} \cup \partial A$ であることが 知られている.
* $A$の任意の点は自身を含む開集合で$A$と自明に接しているので閉包の点である．つまり，$A\subseteq \overline{A}$

**集積点（極限点）と閉包**

$x \in X$が$A$の集積点（極限）であるとは，$x$の任意の開近傍$U(x)$が$x$を除いて$A$と交わること

$$(U(x) \backslash\{x\}) \cap A \neq \varnothing$$

* この定義から，$A$の集積点は$A$に含まれている必要はありません．
* $A$の集積点は閉包 $\bar{A}$ に含まれます．集䅡点のすべての開近傍は $A$ と交わるためです.
* 集積点は，そのどんな小さい開近傍を取ってきても $A$ と交わるので，集積点の周りには常に$A$の点が存在します.

**閉包は閉集合**

閉包は閉集合です．証明しましょう．

**証明**

閉包の補集合が開集合であることを示します．

$x \in \bar{A}^c=X \backslash \bar{A}$ なる閉包の補集合の要素を考えます．定義より，$x$は触点ではありません．触点の定義の否定から, $x$ のある開近傍 $U(x)$ が存在して, $U(x) \cap A=\varnothing$ となる. すなわち, ${ }^{\forall} y \in U(x), y \notin \bar{A}$ である. なぜなら， $y \in \bar{A} \Rightarrow U(x) \cap A \neq \varnothing$ となり矛盾する.

${ }^{\forall} y \in U(x), y \in \bar{A}^c$ より $U(x) \subseteq \bar{A}^c$, すなわち, ${ }^{\forall} x \in \bar{A}^c,{ }^{\exists} U(x)$ s.t $U(x) \subseteq \bar{A}^c$ となり, $\bar{A}^c$ の任意の点は内点であるので, $\bar{A}^c$ は内点の集合なので開集合である.


## 位相空間における閉包と稠密性

閉包は稠密性の議論に使います．
$A$を位相空間$X$の部分集合とします．

* $A$が稠密であるとは，$A$の閉包$\bar{A}$について$\bar{A}=X$が成立することを言います．
  * つまり，$A$の閉包が$X$を覆っています
  * つまり，$x \in X$の近傍には$a \in A$がいます．
  * つまり，$X$のどんな点も$A$の点の極限として捉えることができます．
    * 例えば実数$\mathbb{R}$に対して有理数$\mathbb{Q} \subset \mathbb{R}$は稠密です．
