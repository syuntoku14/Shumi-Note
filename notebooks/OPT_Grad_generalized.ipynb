## 一般微分可能関数と疑似勾配

---

**一般微分可能関数**

関数$f: \mathbb{R}^n \to \mathbb{R}^1$は次を満たすときに$x \in \mathbb{R}^n$で一般微分可能であるという．

$x$の何らかの近傍において，次を満たすような$x$について上半連続な集合値関数$G_f(y) (y \in \mathbb{R}^n)$が定義される：

* $G_f(y)$は空でなく，有界で，凸な閉集合である．
* $x$の近傍の$y$で次が成立する．
$$
f(y)=f(x)+\langle g, y-x\rangle+o(x, y, g)
$$
ここで$g \in G_f(y)$であり，$o(x, y, g)$は任意の系列$y^k\to x$，$g^k \in G_f(y^k)$について次を満たす：
$$
\lim _{k \rightarrow \infty}\frac{o\left(x, y^k, g^k\right)}{\left\|y^k-x\right\|}=0
$$

また，$g \in G_f(y)$は$f$の$y$における疑似勾配と呼ぶ．

ちなみに，集合値関数$f$が次を満たすときに上半連続といいます：

任意の$x_0 \in X$について，系列$\left\{x_n\right\}$が$x_0$に収束し，任意の$y_n \in F\left(x_n\right)$なる系列$\left\{y_n\right\}$が，$\left(y_n, F\left(x_0\right)\right) \rightarrow 0, n \rightarrow \infty$を満たす．

---

この定義は実際にいろいろな微分の一般化になってます．
次が成立します．

---

**定理**: 連続微分可能（$C^1$級）な関数$f$は一般微分可能であり，その勾配は疑似勾配である．

**証明**

$g(y)$を$f(y)$の勾配とします．このとき，
$$
o(x, y, g)=\langle g(x)-g(y), y-x\rangle+o(\|y-x\|)
$$
とすると，
$$
\begin{aligned}
f(y) & =f(x)+\langle g(x), y-x\rangle+o(\|y-x\|) \\
& =f(x)+\langle g(y), y-x\rangle+o(x, y, g),
\end{aligned}
$$
が成立します．

この$o(x, y, g)$は明らかに上で定義した条件を満たします．

---

**定理**: [Weakly convexな関数](CVX_weakly_convex_and_Moreau_envelope.ipynb)も一般微分可能であり，その劣勾配は疑似勾配になる．
（論文中ではもうちょっと一般的なweakly convexの定義を導入してるが，簡単なので２ノルムでやる）

**証明**

$f$をweakly convexな関数とすると，任意の$g \in \partial f(x)$と $x, y \in \mathcal{X}$について，

$$
f\left(y\right)\geq f(x)  + \left\langle g, y-x \right\rangle-\frac{\ell}{2}\left\|x-y\right\|^2
$$

が成立します．
ここで，
$$
o(x, y, g)=f(y)-f(x)-\langle g, y-x\rangle
$$
とします．
$g(x) \in \partial f(x)$, $g(y) \in \partial f(y)$とかくと， 
$$
\begin{aligned}
&\langle g(x)-g(y), y-x\rangle 
-\frac{\ell}{2}\left\|x-y\right\|^2\\
=& o(x, y, g(y)) - o(x, y, g(x))
-\frac{\ell}{2}\left\|x-y\right\|^2\\
\leq & o(x, y, g(y)) \leq \frac{\ell}{2}\left\|x-y\right\|^2
\end{aligned}
$$

ここで，
$$
\begin{aligned}
&-o(x, y, g(x))-\frac{\ell}{2}\left\|x-y\right\|^2\\
=&-f(y)+f(x)+\langle g(x), y-x\rangle -\frac{\ell}{2}\left\|x-y\right\|^2
\leq 9
\end{aligned}
$$
と
$$
\begin{aligned}
o(x, y, g(y)) &= f(y)-f(x)-\langle g(y), y-x\rangle\\
&= - (f(x) - f(y) -\langle g(y), x - y\rangle) 
\leq \frac{\ell}{2}\left\|x-y\right\|^2
\end{aligned}
$$

ここで，$g(y)$を一番$g(x)$に近くなるように選びましょう：
$$
\|g(x)-g(y)\|=\rho(g(y), \partial f(x)) \equiv \inf _g\{\|g(y)-g\|: g \in \partial f (x)\}
$$
とします．このとき，コーシー・シュワルツより
$$
-\rho(g(y), \partial f(x))\|y-x\| \leq o(x, y, g(y)) \leq \frac{\ell}{2}\left\|x-y\right\|^2
$$

ここで，$\|y - x\| \to 0$とすると，
$$
-\rho(g(y), \partial f(x)) \leq \frac{o(x, y, g(y))}{\|y - x\|} \leq \frac{\ell}{2}\left\|x-y\right\|
$$
は左右が０に行くので，$o(x, y, g)$は条件を満たします．

## Pointwise maximumと疑似勾配

もともとの目的を示しましょう．次が成立します．

---

**定理**

$f_i(y) \quad\left(y \in \mathbb{R}^n, i=1, \ldots, m\right)$を一般微分可能な関数の集合とします．
$G_{f_i}(y)$はその疑似勾配の写像です．このとき，
$$
f(y)=\max \left(f_1(y), \ldots, f_m(y)\right)
$$
であり，
$$
G_f(y)=\operatorname{co}\left\{g \in \mathbb{R}^n \mid g \in G_{f_l}(y), f_i(y)=f(y)\right\}
$$
が成り立ちます．

**証明**

$m=2$のときを証明すれば十分です．つまり，上で定義した$G_f(y)$が，ちゃんと疑似勾配の定義と同じになってることを示せば良いです．

**上半連続の証明**

まずは$G_f$がClosedかつlocally boundedであることを示します．これにより，上半連続であることが言えます．
Closedであるとは，つまり$y^k \to y$，$g^k \in G_f(y^k)$，$g^k \to g$ならば，$g\in G_f(y)$ であることを言います．

ここで，
$$
g^k=\sum_{i=1}^m \lambda_i^k g_l^k, \quad \lambda_i^k \geqslant 0, \quad \sum_{i=1}^m \lambda_i^k=1
$$
としましょう．$f_i\left(y^k\right)<f\left(y^k\right)$ならば$\lambda_i^k=0$とします．
$\lambda^k_i$がバウンドされており，さらに$G_{f_i}$がclosedかつlocally boundedであり，そして$m$が有限なので，
$$
\lambda_i^{k_s} \rightarrow \lambda_i, \quad g_i^{k_s} \rightarrow g_i, \quad s=0,1, \ldots
$$
な部分列が存在し， ．．．
