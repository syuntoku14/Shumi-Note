# 大数の法則について

参考：
* [The Laws of Large Numbers Compared](http://www.dklevine.com/archive/strong-law.pdf)

準備：
* $X_1, X_2, \dots$を独立な確率変数とします．それぞれの期待値を$\mu_i$とします．
* $\bar{X}_n=\frac{1}{n} \sum_{i=1}^n X_i$
* $\bar{\mu}_n=\frac{1}{n} \sum_{i=1}^n \mu_i$
* $\mathbb{E}[X_n^2] < \infty$のとき，$X_n \in L^2$ と書きます．ちなみに$X_n \in L^2$ならば$\mathbb{E}[|X_n|] < \infty$です．（[これなど](https://homepage.univie.ac.at/erhard.reschenhofer/pdf/advanced/L2.pdf)）

大数の法則には大きく弱法則と強法則の２種類があります．ちゃんと見てみましょう．

## 大数の弱法則

次の定理を大数の弱法則といいます．

---

独立な$(X_n)_{n=1}^\infty$が$X_n \in L^2$かつ
$$
\mathbb{V}[X_n] = \mathbb{E}[(X_n - \mu_n)^2] = \sigma_n^2
$$
が
$$
\sup_n \frac{\sum^n_{k=1} \sigma_k^2}{n} < \infty
$$
をみたすとする（例えば分散が上から抑えられていればこれは成り立つ）．このとき，

$$
\forall_{\varepsilon>0} \lim _{n \rightarrow \infty} \operatorname{Pr}\left(\left|\bar{X}_n-\bar{\mu}_n\right| \leq \varepsilon\right)=1
$$

---

このとき，**$\bar{X}_n$は$\bar{\mu}_n$に確率収束する**といいます．
丁寧に書くと，$\forall \varepsilon$について
$$
\lim_{n \to \infty}
\operatorname{Pr}\left(\{\omega \in \Omega \mid |X_n(\omega) - X(\omega)| > \varepsilon\}\right) = 0
$$
のとき，$X_n$は$X$に確率収束する，といいます．

また，ε-δ論法を使えば

$$
\forall_{\varepsilon>0} \forall_{\delta>0} \exists_{N>0} \forall_{n \geq N} \operatorname{Pr}\left(\left|\bar{X}_n-\bar{\mu}_n\right| \leq \varepsilon\right) \geq 1-\delta
$$
と書くこともできます．

**証明**

独立性より，$i \neq j$のとき，
$$
\mathbb{E}[(X_i - \mu_i)(X_j - \mu_j)] = \mathbb{E}[X_i]\mathbb{E}[X_j] - \mu_i \mu_j = 0
$$
よって，
$$
\begin{aligned}
\mathbb{E}[(\bar{X}_n - \bar{\mu}_n)^2]
&= \frac{1}{n^2} \mathbb{E}\left[\left(\sum^n_{i=1} X_i - \mu_i\right)^2\right]\\
&= \frac{1}{n^2} \sum^n_{i=1}\mathbb{E}\left[\left(X_i - \mu_i\right)^2\right]\\
&= \frac{1}{n} \frac{\sum^n_{i=1}\sigma^2_i}{n}
\end{aligned}
$$
仮定より二項目は$< \infty$なので，$n\to \infty$でこれは$0$に収束します．

これは$p$次平均収束そのものです．

---

**$p$次平均収束**

$p > 0$とする．$L^p \triangleq \{確率変数 X \mid |X|^pが可積分\}$という表記を導入する．

このとき，
$X_n, X\in L^p$ かつ$\lim_{n\to \infty} \mathbb{E}[|X_n - X|^p]=0$のとき，$X_n$は$X$に$p$次平均収束するという．

実は$p$次平均収束するならば確率収束します．
証明はチェビシェフの不等式を使えば簡単です．実際，$\forall \varepsilon > 0$に対して，

$$
\mathbb{P}(|X_n - X| \geq \varepsilon)
= \mathbb{E}(1\{|X_n - X| \geq \varepsilon\})
\leq \frac{\mathbb{E}|X_n - X|^p}{\varepsilon^p} \to 0
$$
ここで不等式はチェビシェフの不等式を使ってます（[一般版](https://en.wikipedia.org/wiki/Chebyshev%27s_inequality)）

---


## 中心極限定理

弱法則では，$\sigma^2_n$が有界な場合は成立し，$\bar{X}_n$が$\mu$に確率収束することが言えます．
一方で，有界ならば中心極限定理が使え，もっと強いことが言えます．

---

**中心極限定理**

$(X_n)_{n=1}^\infty$が平均$\mu$，分散$\sigma^2$のi.i.d.の確率変数とする．
このとき，
$$
\frac{\sqrt{n}}{\sigma}\left(\bar{X}_n - \mu\right) \rightsquigarrow \mathcal{N}(0, 1)
$$
が成立する．

## 大数の強法則

$(X_n)_{n=1}^\infty$はi.i.d.であり，$X_1 \in L^4$とします．このとき，

$$
\operatorname{Pr}\left(\lim _{n \rightarrow \infty} \bar{X}_n=\mu\right)=1
$$
が成立します．

このとき，**$\bar{X}_n$が$X$に概収束する**といいます．


丁寧に書くと，
$$
\operatorname{Pr}\left(\{w \in \Omega \mid \lim_{n \to \infty} X_n(\omega) = X(\omega)\}\right) = 1
$$
のとき，$X_n$は$X$に概収束するといいます．
