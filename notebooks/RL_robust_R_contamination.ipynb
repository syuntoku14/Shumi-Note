# ロバストMDPでのR-contamination setは割引率変えただけ

* [Policy Gradient in Robust MDPs with Global Convergence Guarantee](https://arxiv.org/abs/2212.10439)

ロバストMDPだとしばしばR-contaminationモデルってのが出てきます．$(s, a)$-rectangular 集合について，$\mathcal{P}=\underset{s \in \mathcal{S}, a \in \mathcal{A}}{\times} \mathcal{P}_{s, a}$が次を満たす集合をR-contaminationと呼びます．

$$
\mathcal{P}_{s, a}:=\left\{(1-R) \overline{\boldsymbol{p}}_{s a}+R \boldsymbol{q} \mid \boldsymbol{q} \in \Delta(S)\right\}, s \in \mathcal{S}, a \in \mathcal{A} .
$$

R-contamination modelでのベルマン作用素は次で定まります：

$$
\begin{aligned}
\left(\mathcal{T}^r \boldsymbol{v}\right)_s: & =\min _{a \in \mathcal{A}} \max _{\boldsymbol{p}_{s a} \in \mathcal{P}_{s, a}}\left(c_{s a}+\gamma \boldsymbol{p}_{s a}^{\top} \boldsymbol{v}\right) \\
& =\min _{a \in \mathcal{A}} c_{s a}+\gamma\left[(1-R) \overline{\boldsymbol{p}}_{s a}^{\top} \boldsymbol{v}+R \max _{s^{\prime}} v_{s^{\prime}}\right] \\
& =\left[\min _{a \in \mathcal{A}} c_{s a}+\gamma(1-R) \overline{\boldsymbol{p}}_{s a}^{\top} \boldsymbol{v}\right]+R \gamma \max _{s^{\prime}} v_{s^{\prime}}\\
& =(\mathcal{T} \boldsymbol{v})_s+R \gamma\|\boldsymbol{v}\|_{\infty}
\end{aligned}
$$

ここで，$\mathcal{T}$はR-contaminationの中心を遷移，割引率が$\gamma(1-R)$についての普通のベルマン作用素です：
$$
(\mathcal{T} \boldsymbol{v})_s:=\min _{a \in \mathcal{A}} c_{s a}+\gamma(1-R) \overline{\boldsymbol{p}}_{s a}^{\top} \boldsymbol{v}
$$

このとき，
$$
\begin{aligned}
\mathcal{T}^r \boldsymbol{v}^r & =\mathcal{T} \boldsymbol{v}^r+R \gamma\left\|\boldsymbol{v}^r\right\|_{\infty} \boldsymbol{e} \\
\Longleftrightarrow\left(\mathcal{T}^r\right)^2 \boldsymbol{v}^r & =(\mathcal{T})^2 \boldsymbol{v}^r+R \gamma\left\|\mathcal{T}^r \boldsymbol{v}^r\right\|_{\infty} \boldsymbol{e}+R \gamma^2(1-R)\left\|\boldsymbol{v}^r\right\|_{\infty} \boldsymbol{e} \\
& =(\mathcal{T})^2 \boldsymbol{v}^r+\left[R \gamma+R \gamma^2(1-R)\right] \cdot\left\|\boldsymbol{v}^r\right\|_{\infty} \cdot \boldsymbol{e} \\
\Longleftrightarrow\left(\mathcal{T}^r\right)^k \boldsymbol{v}^r & =(\mathcal{T})^k \boldsymbol{v}^r+\left[R \gamma+R \gamma^2(1-R)+R \gamma^3(1-R)^2+\cdots\right] \cdot\left\|\boldsymbol{v}^r\right\|_{\infty} \cdot \boldsymbol{e} \\
& =(\mathcal{T})^k \boldsymbol{v}^r+\sum_{n=1}^k R \gamma^k(1-R)^{k-1} \cdot\left\|\boldsymbol{v}^r\right\|_{\infty} \cdot \boldsymbol{e} .
\end{aligned}
$$

両辺で極限を取ると（元論文では最後の分子の$R\gamma$が抜けてるので間違い．これがないと$v^{nr}$が負になっちゃう），
$$
\begin{aligned}
\lim _{k \rightarrow \infty}\left(\mathcal{T}^r\right)^k \boldsymbol{v}^r & =\boldsymbol{v}^r \\
& =\lim _{k \rightarrow \infty}\left[(\mathcal{T})^k \boldsymbol{v}^r+\sum_{n=1}^k R \gamma^n(1-R)^{n-1} \cdot\left\|\boldsymbol{v}^r\right\|_{\infty} \cdot \boldsymbol{e}\right] \\
& =\boldsymbol{v}^{n r}+\lim _{k \rightarrow \infty}\left[\sum_{n=1}^k R \gamma (\gamma (1-R))^{n-1} \cdot\left\|\boldsymbol{v}^r\right\|_{\infty} \cdot \boldsymbol{e}\right] \\
& =\boldsymbol{v}^{n r}+\frac{R\gamma}{1-\gamma(1-R)} \cdot\left\|\boldsymbol{v}^r\right\|_{\infty} \cdot \boldsymbol{e} .
\end{aligned}
$$
が得られます．

よって，
$
v^r = v^{nr} + \frac{R\gamma}{1-\gamma+R\gamma} \cdot\left\|\boldsymbol{v}^r\right\|_{\infty} \cdot \boldsymbol{e}
$
なので，$R$-contaminationモデルの最適価値は割引率が$\gamma (1-R)$の価値に定数を足しただけです．
以上から，$R$-contaminationモデルの最適方策は割引率が$\gamma (1-R)$のMDPの最適方策と同じです．



**補足**

$$
\frac{R\gamma}{1-\gamma(1-R)} 
= \frac{R\gamma}{1-\gamma+R\gamma} 
= \frac{1}{1 + \frac{1-\gamma}{R\gamma}} 
=: \rho \leq 1
$$
でした．

更に変形すると，
$$
\begin{aligned}
\left\|\boldsymbol{v}^r\right\|_{\infty}
& =
\left\|
\boldsymbol{v}^{n r}
\right\|_\infty
+\rho\left\|\boldsymbol{v}^{nr}\right\|_{\infty}
= 
\left\|
\boldsymbol{v}^{n r}
\right\|_\infty
+\rho\left\|\boldsymbol{v}^{nr}\right\|_{\infty}
+\rho^2\left\|\boldsymbol{v}^{nr}\right\|_{\infty}
\dots = 
\frac{1}{1-\rho} 
\left\|
\boldsymbol{v}^{n r}
\right\|_\infty
\end{aligned}
$$
です．
ちなみに

$$
\frac{1}{1-\rho}
=
\frac{1}{1-\frac{1}{1 + \frac{1-\gamma}{R\gamma}}} 
=
\frac{1 + \frac{1-\gamma}{R\gamma}}{1 + \frac{1-\gamma}{R\gamma}-1} 
=
\frac{R\gamma + 1-\gamma}{1-\gamma} 
$$

なので，結局

$$
v^r = v^{nr} + \frac{R\gamma}{1-\gamma} \cdot\left\|\boldsymbol{v}^{nr}\right\|_{\infty} \cdot \boldsymbol{e}
$$

になります．
