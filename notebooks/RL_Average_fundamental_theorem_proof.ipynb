# マルコフ連鎖の基本定理の証明

参考：
* [Various proofs of the Fundamental Theorem of Markov Chains](https://arxiv.org/abs/2204.00784)
* [鈴木先生のマルコフ連鎖](https://ibis.t.u-tokyo.ac.jp/suzuki/lecture/2019/probth/PME_2_ver3.pdf)

エルゴード性があるマルコフ連鎖では，定常分布が存在し，その定常分布が連鎖の極限で得られることが知られています（ここで，初期分布には依存しません）．

この事実のことを「基本定理（fundamental theorem）」と言います．
基本定理は非常に重要な定理で，マルコフ連鎖の応用の根幹を担ってます．

例えばインターネットのpage rank algorithmが最も有名な応用事例でしょう．詳細は論文に任せますが，web searchをマルコフ連鎖としてモデル化して，定常分布の確率が高い順に表示することで推薦を行ってるっぽいです．

今回は基本定理の様々な証明を学びます．

**表記**

* $\Omega$：状態空間
* $\Omega \times \Omega$の行列$P$の$P(x, \cdot)$で$P$の$x$行目を表す．これは$x$から次状態に行く遷移確率
* 定常分布：$\pi P = \pi$を満たす$\pi$のこと
* 任意の要素$i, j$で$A_{ij} > 0$の行列を正と呼ぶ．
* $G(P)$：$P$の背後にあるグラフ．
  * $G(P)$が[強連結](https://manabitimes.jp/math/1250)のとき，つまりすべてのノードに任意のノードからアクセス可能なとき，$G(P)$を既約である，という．
  * 状態$s$について，$s$からスタートして$s$に戻ってくるのが$d(s)$の倍数に限るとき，$d(s)$のことを連鎖の周期と言います．任意の状態について$d(s)=1$のとき，$G(P)$を非周期的といいます．
* 一般に，$i\to j$かつ$j \to i$のとき，$i, j$は$P$に関して**同値**である，といいます．communicatingのことです．


## 周期についての定理

### 定理：同値類の状態の周期は同じ

$i, j \in \Omega$をrecurrentな状態とします．
$i \leftrightarrow j$ならば，$d(i)=d(j)$です．

**証明**
まず，$i \leftrightarrow j$なので，
$P^k_{i,j} > 0$かつ$P^\ell_{i,j} > 0$を満たす$k, \ell$が存在します．
このとき，
$$
P^{k+\ell}_{i, i} \geq P^{k}_{i, j}P^{\ell}_{j, i} > 0
$$
* なので，$k+\ell$ステップで$i$から$i$に帰ってくれるので，$k+\ell$は$d(i)$の倍数です．

一方で．$P^m_{j, j} > 0$である任意の$m$について，
$$
P^{k+\ell+m}_{i, i} \geq P^{k}_{i, j}P^m_{j, j}P^{\ell}_{j, i} > 0
$$
が成り立ってます．
* よって，$k+\ell+m$も$d(i)$の倍数になります．
* ここで，$k+\ell$が$d(i)$の倍数なので，$m$も$d(i)$の倍数になります．
* ところで，$d(j)$は以上を満たす$m$の最大公約数なので，$d(i)\times 何か=m$は$d(j)$以下になります．

以上の$j$と$i$を入れ替えても同じことが言えるので，$d(i) \geq d(j)$も言えます．
よって，$d(i)=d(j)$です．


### 定理：$i\leftrightarrow j$のときの往路の性質

$i\leftrightarrow j$のとき，その共通の周期を$d$とすると，
$P^{\ell}_{i, j}>0, P^{m}_{i, j}>0 \Rightarrow \ell=m \bmod d$
が成り立ちます．

**証明**
$j\to i$なので，ある$n$が存在して，$P^n_{j, i} > 0$です．よって，
$$
\begin{aligned}
& P^{\ell+n}_{i, i} \geq P^{\ell}_{i, j} P^{n}_{j, i}>0 \\
& P^{m+n}_{i, i} \geq P^{m}_{i, j} P^{n}_{j, i}>0
\end{aligned}
$$
です．よって，$\ell+n$と$m+n$回の移動で$i\to i$ができるので，どちらも$d$の倍数です．

よって，$\ell - m = (\ell + n) - (m+n)$も$d$の倍数になるので，$\ell = m \operatorname{mod} d$です．

---

以上の定理を使うと，既約かつ閉なクラス$R \subset \Omega$を次のように分割できます：

$R$内の状態$i$を基準として，各$j \in R$を考えましょう．
そして，$P^\ell_{i, j} > 0$なる$\ell$を見つけたとします．このとき，

$$
R=C_i(1) \cup C_i(2) \cup C_i(3) \cup \cdots \cup C_i(d)
$$

として分割できます．ここで，$C_i$には$\ell$を$d$で割った余りが$i$のものを割り当ててます．


### エルゴード性についての定理

* [Various proofs of the Fundamental Theorem of Markov Chains](https://arxiv.org/abs/2204.00784)

### 補題：エルゴードなMDPではいつか正の確率行列になる

有限マルコフ連鎖の$P$がエルゴードであるとき，つまり既約かつ非周期的ならば，任意の$l \geq k$について$P^l$が正になるような$k$が存在する．

**証明**

まず，$i \in \Omega$について，
$$
A_{i i}=\{|w| \mid w \text { is a walk from } i \text { to } i,|w| \geq 1\}
$$
を$i$から$i$に戻るための経路の長さの全パターンとします．
連鎖が非周期的なので，$A_{ii}$の最大公約数は$1$です．

ここで，$A_{ii}$が和について閉じていることを意識しておきましょう．
実際，$i\to i$の経路 $w_1$と$w_2$がある場合，$|w_1|, |w_2| \in A_{ii}$であり，さらに$w_1$と$w_2$を組み合わせても$i\to i$なので，$|w_1| + |w_2| \in A_{ii}$でもあります．

さらに話を進めるために，次を使います：

---

$A=\{a_1, a_2, \dots\}$を次を満たす正の整数の列とします．

1. $\operatorname{gcd}\left\{a_1, a_2, \ldots\right\}=1$
2. 和について閉じている

このとき，ある$N < \infty$が存在し，任意の$n \geq N$について$n \in A$が成り立ちます．

---

上の事実を$A_{ii}$に使うと，ある$N_i < \infty$が存在して，任意の$n \geq N_i$について$i\to i$に$n$ステップで帰ることができます．

さて，$j$を連鎖$M$上の任意の状態とします．$M$は既約なので，$G(P)$上には$i\to j$に行く経路が少なくとも１つ存在します．その長さを$l_{ij}$としましょう．

ここで，$K_{ij}=N_i + l_{ij}$としましょう．
上の事実から，任意の$m \geq K_{ij}$について，長さ$m$で$i\to j$に行く経路が存在します（最初の$m - l_{ij}$の長さを$i\to i$で消費すれば良い）．

よって，ある$K$が存在して，任意の$l \geq K$で$P^l$が正になります．

---


## 基本定理とその証明：簡単なやり方

* $P^{\infty} \stackrel{\text { def }}{=} \lim _{n \rightarrow \infty} P^n$とします．

このとき，エルゴード性があるときは次が成り立ちます．

1. （列が定数）：任意の正な$n\times n$の確率行列$P$について，次を満たす$\pi_1, \pi_2, \ldots, \pi_n$が存在します：

$$
P^{\infty}=\left[\begin{array}{cccc}
\pi_1 & \pi_2 & \ldots & \pi_n \\
\pi_1 & \pi_2 & \ldots & \pi_n \\
\cdot & \cdot & \ldots & \cdot \\
\pi_1 & \pi_2 & \ldots & \pi_n
\end{array}\right]
$$
2. （定常分布）：そして，$\pi \stackrel{\text { def }}{=}\left(\pi_1, \pi_2, \ldots, \pi_n\right)$は$P$の定常分布です．
3. （任意の初期状態）：また，任意の初期状態分布$\sigma=\left(\sigma_1, \sigma_2, \ldots, \sigma_n\right)$からスタートした系列$\left(\sigma^{(i)}\right)_{i \geq 1}$は，$i\to \infty$で$\pi$に収束します．


ここで，**既約性**と**非周期性**の両方が基本定理の必要条件であることに注意しましょう．
1. もし連鎖が２つのつながっていない要素$A, B$からなる場合は，初期状態が$A$ or $B$で定常分布が異なります．よって，既約性がないと，任意の初期状態で．．．の主張を実現できなくなります
2. 非周期的な場合に定常分布に収束しないケースは
$
P=\left[\begin{array}{ll}
0 & 1 \\
1 & 0
\end{array}\right]
$
において，初期状態をどちらかの状態に固定したケースを考えればすぐにわかります．この例では定常分布自体は存在することに注意しましょう．ただ，定常分布に収束しません．




これを示すために，次を示します：

### 補題：確率行列の収束

$P$を正な確率行列ととする．
$P^i$について，任意の固定された列$k$について，その最大と最小値を$m^i$と$M^i$とすると，以下が成立します：
1. $(m^i)_{i \geq 1}$は非減少
2. $(M^i)_{i \geq 1}$は非増加
3. $\Delta^i = M^i - m^i$は$i\to \infty$で0に収束する

この補題から，次が直ちに成立します：

* $P$が正の確率行列のとき，$P^i$の$k$列目が$i\to \infty$で$\pi_k$に収束するような値$\pi_k$が存在する
    * 証明：$P^i$も正の確率行列なので，$0 < m^i \leq M^i < 1$です．任意の有界かつ非減少（or 非増加）な系列は極限を持ちます．また，3つ目の性質から，１と２の極限は同じになります（つまり，任意の$\epsilon > 0$について，$\Delta^i < \epsilon$）．
    * ちなみに３つ目単体では不十分です．例えば$((-1)^i)_i$なる系列は収束しないですが，２つの系列がどちらもこれに従うとき，その差が０になってます．

さて，この確率行列の収束性を証明してみましょう．

---

**1と2の証明**

まず，$m$の定義から，
$$
\begin{aligned}
m^{(i+1)} & \stackrel{\text { def }}{=} \min _r P_{r k}^{i+1} \\
& =\min _r \sum_s P_{r s} P_{s k}^i \\
& \geq \min _r \sum_s P_{r s} m^{(i)} \\
& =m^{(i)} \min _r \sum_s P_{r s} \\
& =m^{(i)}
\end{aligned}
$$
なので，$m^{i+1} \geq m^i$です．同様にして，$M^{i+1}\leq M^i$も言えます．

**3の証明**

さて，$P^{i+1}$の，任意の$r, k$番目の要素を考えましょう．
$P^{i+1} = PP^i$なので，$P_{r k}^{i+1}=\sum_l P_{r l} P_{l k}^i$です．
$P^i$の$k$列目の最大値$M^i$が$s$行目にあるとします．このとき，

$$
\begin{aligned}
P_{r k}^{i+1} & =P_{r s} M^{(i)}+\sum_{l \neq s} P_{r l} P_{l k}^i \\
& \geq P_{r s} M^{(i)}+\left(1-P_{r s}\right) m^{(i)} \\
& =m^{(i)}+P_{r s}\left(M^{(i)}-m^{(i)}\right) \\
& \geq m^{(i)}+p_{\min }\left(M^{(i)}-m^{(i)}\right)
\end{aligned}
$$

ここで，$p_{\min}$は$P$の中で一番小さい値を指します．
これは任意の$r, k$について成り立つので，もちろん$m^{i+1}$を与える要素でも成り立ちます．よって，
$$
m^{(i+1)} \geq m^{(i)}+p_{\min }\left(M^{(i)}-m^{(i)}\right)
$$
が成り立ちます．同様にすると，
$$
M^{(i+1)} \leq M^{(i)}-p_{\min }\left(M^{(i)}-m^{(i)}\right)
$$
も成立します．２つを合体させると：
$$
M^{(i+1)}-m^{(i+1)} \leq\left(1-2 p_{\min }\right)\left(M^{(i)}-m^{(i)}\right)
$$
であり，$\Delta^{(i+1)} \leq\left(1-2 p_{\min }\right) \Delta^{(i)}$が得られます．$\Delta^{(1)} \leq 1$なので，繰り返すと
$$
\Delta^{(n)} \leq\left(1-2 p_{\min }\right)^{n-1}
$$
が得られます．ここで，マルコフ連鎖が３以上の状態を持つとき，明らかに
$0<\left(1-2 p_{\min }\right)<1$なので，上の式から$\Delta^{(n)}\to 0$です．
よって，
$$
P^\infty=
\left[\begin{array}{cccc}
\pi_1 & \pi_2 & \ldots & \pi_n \\
\pi_1 & \pi_2 & \ldots & \pi_n \\
\cdot & \cdot & \ldots & \cdot \\
\pi_1 & \pi_2 & \ldots & \pi_n
\end{array}\right]
$$
が成立します．

---

* この$\pi$が定常分布になってる
* 任意の初期分布$\sigma$から始まっても定常分布に収束する

ことは論文参照．

### （補足）混合時間について

有限マルコフ連鎖は[#P-困難](https://ja.wikipedia.org/wiki/%EF%BC%83P)な問題や組み合わせ最適化の近似解を得るために使われます．
このとき，アルゴリズムの効率性はマルコフ連鎖が定常分布に収束する速さと密接に関連しています．

この，定常分布に収束する速さのことを混合時間と呼びます．
上でやった基本定理の証明を使うと，混合時間の上界を得ることができます．

混合時間はTV距離を使って表現します：
* $\|\mu-\nu\|_{\mathrm{TV}} \stackrel{\text { def }}{=} \frac{1}{2} \sum_{x \in \Omega}|\mu(x)-\nu(x)|$

$t$時刻経過したマルコフ連鎖の，定常分布$\pi$からの距離$d(t)$を次のように定義しましょう：

$$
d(t) \stackrel{\text { def }}{=} \max _{x \in \Omega}\left\|P^t(x, .)-\pi\right\|_{\mathrm{TV}}
$$

このとき，$\epsilon$が与えられたときの混合時間は次で定義されます：

$$
t_{\operatorname{mix}}(\epsilon) \stackrel{\text { def }}{=} \min \{t \mid d(t) \leq \epsilon\}
$$

（ちなみにMixing timeは多くの場合で$|epsilon = 1/4$のときを指すことがあります）

さて，基本定理の証明を思い出しましょう．列$j$に注目して，任意の行$i$での定常分布からの差は$\left|P_{i j}^t-\pi_j\right| \leq \Delta^{(t)}$で表現されます．

これは任意の行（状態）$i$で成立するので，TV距離の定義から，$d(t) \leq n \Delta^{(t)}$が成立します．

さらに，$\Delta^{(t)} \leq\left(1-2 p_{\min }\right)^{t-1}$なので，混合時間については
$$
\begin{gathered}
t \geq \frac{1}{2 p_{\min }} \ln \frac{n}{\epsilon}+1 \Rightarrow d(t) \leq \epsilon \\
t_{\operatorname{mix}}(\epsilon)=O\left(\frac{1}{2 p_{\min }} \ln \frac{n}{\epsilon}\right)
\end{gathered}
$$

が言えます．

**一般のエルゴードなマルコフ連鎖について**

この連鎖はすべての$P > 0$を仮定しましたが，一般のエルゴードでも同じようなことが言えます．上で示したように，エルゴードな連鎖$Q$では，$m \stackrel{\text { def }}{=} \min \left\{k \geq 1: Q^k>0\right\}$であるような有限な$m$が存在します．$m$以降ではずっと遷移が正なわけですね．
これを使うと，一般の$Q$では
$$
t_{\operatorname{mix}}(\epsilon)=O\left(\frac{m}{2 p_{\min }} \ln \frac{n}{\epsilon}\right)
$$
が言えます．

---

一般に，「マルコフ連鎖を説明する長さ」について混合時間が多項式時間であるとき，早い混合時間であると言います．
ここで，説明する長さは状態の数でも良いですが，一般に状態の数は指数的です．
この辺の議論は12ページ末を参照してください．


## 定常分布の存在と収束を別々に証明するやり方

基本定理の証明のやり方は一つだけではないです．
次を別々に証明しましょう：

1. 連鎖が既約なとき，全状態で非ゼロの確率を持った定常分布が存在する
2. $\pi$がエルゴードな連鎖$P$の定常分布のとき，任意の初期分布からスタートしても，連鎖は$\pi$に収束する

これは唯一性については言ってないですが，唯一性は２つ目 or 既約性から言えます．

### 任意の初期状態 → 唯一性の証明

$\pi, \pi'$が２つの定常分布としましょう．$\pi$を初期分布とします．このとき，$\pi P^n$は$n\to \infty$で$\pi'$に収束します．
しかし，$\pi=\pi P$なので，$n\to \infty$で$\pi$に収束します．よって，$\pi=\pi'$です．

---

明らかに，１と２を証明すると，基本定理が示されます．
１つ目の補題は「既約性が定常分布の状態の十分条件である」ことを表してますが，**既約であることは，必要条件ではありません**．

* 例えば$P=I$の単位行列を考えると，明らかにこれは既約ではない（全連結ではない）ですが，定常分布が存在します．よって，定常分布が存在しても，既約とは限りません．
* また，周期的な連鎖も定常分布を持つことがあります．$\{s_0, s_1\}$の連鎖を考えてみましょう．$s_i$にいる場合は，確率$1$で$s_{1-i}$に移るとします．このとき，定常分布は$[0.5, 0.5]$です（初期状態を$0$ or $1$で取ると定常分布に収束しないことには注意しましょう）．

### 既約 $\to$ 唯一

* Markov Chains and Mixing Times の系1.17にあるらしい

実は，既約であれば，唯一の定常分布になることも言えます．

**補題：既約 $\to$ 唯一**

まず，$P$が既約な連鎖ならば，$h$は定数関数であることを示しましょう．

**証明**

関数$h: \Omega \to \mathbb{R}$は，$h$を列ベクトルとして，$h=Ph$を満たすときに，「調和関数である」と呼ぶことにします．

* $\Omega$は有限なので，$h$の最大値を与える$x_0 \in \Omega$が存在します．$h(x_0)=M$とおきましょう．
* ここで，$P(x_0, z) > 0$かつ$h(z) < M$であるような$z \in Omega$があると仮定します．

このとき，$h$が調和であるとすると，
$$
h\left(x_0\right)=
P\left(x_0, \cdot\right) h=
P\left(x_0, z\right) h(z)
+\sum_{w \neq z} P\left(x_0, w\right) h(w)
$$
が成り立ちます．
$P(x_0, z) > 0$なので，２項目は$< M$以下になります．よって，$h(x_0) < M$であり，矛盾します．よって，$h(z) = M$です．
**つまり，$x_0$から$z$に行けるとき，$z$でも$h$は最大値$M$を取ります．**

さらに，既約（全連結）なので，任意の$y \in \Omega$に対して，$x_0$から$y$に行く経路$x_0, x_1, \dots x_{q}=y$が必ず存在します．
これを上の事実に当てはめると，$M= h(x_0) = h(x_1) = \dots h(y)$が成立します．
**これが任意の$y$について成立するので，$h$は定数関数です．**

---

さて，唯一性の証明に戻ります．
* $P$が確率行列なので，$(P - I)$の核は$x=1$です．
* 次元定理より，$\operatorname{rank} (P - I) = \Omega - 1$です．

行列において，行のランクと列のランクは等しいので，$\sigma = \sigma P$を満たすような解の空間の次元は$1$になります．
$\sigma$が確率行列になるように正規化すると，唯一の$\sigma$しか存在しないことがすぐにわかります．



### 存在性の証明

上では存在するならば唯一であることは示しましたが，そもそも存在するかは示してません．存在性を示しましょう．

#### 確率論的なやり方

定常分布の$\pi_y$は，「長期的に見て，$y$にどれくらい訪れるか？」の確率を表してます．すなわち，ある系列において，$y$が訪問される期待回数は$\pi_y$の定数倍になるはずです．

これを踏まえて定常分布を構築することで，存在性を証明していきます．
* $\tilde{\pi}_y \stackrel{\text { def }}{=} \mathbf{E x}_z$ ( number of visits to $y$ before returning to $z$ for the first time)としましょう．つまり，$\tilde{\pi}_y$は，$z$から始まって，　$z$に戻る前に$y$を再訪問する回数の期待値を表します．
* $\tau^+_z = \min \left\{t \geq 1 \mid X_0=z, X_t=z\right\}$とします．つまり，$z$にもどる初めての時刻です．
* $I_t$を，$X_t =y$かつ$\tau^+_z > t$ならば$1$を出す指示変数とします．つまり，$t$より後で$y$に戻るとき，$1$です．

このとき，
$$
\tilde{\pi_y}=\mathbf{E x}\left(\sum_{t=0}^{\infty} I_t\right)=\sum_{t=0}^{\infty} \operatorname{Pr}_z\left(X_t=y, \tau_z^{+}>t\right)
$$
が成り立ちます．

さて，$\tilde{\pi}$が定常である，つまり，任意の$y \in \Omega$について，$\tilde{\pi}_y=\sum_{x \in \Omega} \tilde{\pi}_x P(x, y)$を示しましょう．
定義から，
$$
\sum_{x \in \Omega} \tilde{\pi}_x P(x, y)=\sum_{x \in \Omega} \sum_{t=0}^{\infty} \operatorname{Pr}_z\left(X_t=x, \tau_z^{+}>t\right) P(x, y)
$$
です．よって，ベイズの定理から，
$$
\begin{aligned}
& \mathbf{P r}_z\left(X_t=x, X_{t+1}=y, \tau_z^{+} \geq t+1\right) \\
& =\operatorname{Pr}_z\left(\tau_z^{+} \geq t+1 \mid X_t=x, X_{t+1}=y\right) \mathbf{P r}_z\left(X_t=x, X_{t+1}=y\right) \\
& =\operatorname{Pr}_z\left(\tau_z^{+} \geq t+1 \mid X_t=x\right) \mathbf{P r}_z\left(X_t=x, X_{t+1}=y\right) \\
& =\operatorname{Pr}_z\left(\tau_z^{+} \geq t+1 \mid X_t=x\right) \mathbf{P r}_z\left(X_t=x\right) P(x, y) \\
& =\operatorname{Pr}_z\left(\tau_z^{+} \geq t+1, X_t=x\right) P(x, y)
\end{aligned}
$$
が言えます．これを使って，
$$
\begin{aligned}
& \sum_{x \in \Omega} \tilde{\pi}_x P(x, y) \\
& =\sum_{t=0}^{\infty} \sum_{x \in \Omega} \operatorname{Pr}_z\left(X_t=x, X_{t+1}=y, \tau_z^{+}>t\right) \\
& =\sum_{t=0}^{\infty} \operatorname{Pr}_z\left(X_{t+1}=y, \tau_z^{+} \geq t+1\right) \\
& =\sum_{t=1}^{\infty} \operatorname{Pr}_z\left(X_t=y, \tau_z^{+} \geq t\right)
\end{aligned}
$$

です．ちょっと変形すると，右辺が$\tilde{\pi}_y$と同じであることが言えます．つまり，$\tilde{\pi}P=\tilde{\pi}$が言えてるわけですね．

この$\tilde{\pi}$を正規化して分布に直すと，$\sum_x \tilde{\pi}_x$で割る（つまり$\mathbf{E x}_z\left(\tau_z^{+}\right)$で割るので），
$$
\pi_x=\frac{\tilde{\pi}_x}{\mathbf{E x}_z\left(\tau_z^{+}\right)}
$$
になります．ここで，$z$は任意に取れるので，$z$を$x$で取ると，定義から$\tilde{\pi}_x=1$であり，
$\pi_x=\frac{1}{\mathbf{E x}_x\left(\tau_x^{+}\right)}$です．これは定義から，$\pi_x > 0$であることがわかります．

---

以上の証明で，非周期性は使ってないことに注意しましょう．ただし，既約性は使ってます（非既約ならば，$\tilde{\pi}_x < \infty$が成り立たない可能性があります．）
